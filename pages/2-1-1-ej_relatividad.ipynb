# Ejemplos de relatividad especial

## Viajar hasta el fin del universo

* Supongamos que una viajera quiere llegar a una galaxia que esta a 100 años-luz. Emprende un viaje hasta allí en una nave espacial. Desde el Sistema de Referencia Inercial (SRI) de la tierra de donde parte, pasarán más de 100 años hasta que llegue, debido a que su velocidad $v$ debe ser inferior a la velocidad de la luz $c$. Sin embargo, el trascurso del tiempo en la nave espacial es más lento debido al efecto relativista de dilatación temporal, por lo que cuando el tiempo de la viajera haya trascurrido un año, en el SRI de la tierra habrán transcurrido varios años, dependiendo de la velocidad $v$ y del valor de $\gamma$:

$$
\Delta t = \gamma \Delta t'
$$

```{figure} ../images/relat_11.png
:alt: relat_11.png
:width: 600px
SRI de la tierra y la galaxia.
```

Es decir, **para cuando la viajera haya envejecido un año, en la tierra podrían haber pasado muchos años**. Si $\gamma = 100$, la viajera habrá recorrido 100 años-luz, percibiendo que ha pasado solo algo más de un año.

Desde el Sistema de Referencia Inercial de la nave (*después de que haya alcanzado su velocidad constante de tal modo que $\gamma = 100$*) es la distancia entre la tierra y la galaxia la que parece haberse recortado. Desde su SRI, son la tierra y la galaxia las que se desplazan. **Desde la tierra parece que la galaxia está a 100 años-luz, mientras que desde la nave parece estar a tan solo un año-luz, debido al efecto relativista de la contracción espacial**.

$$
\Delta x = \frac{\Delta x'}{\gamma}
$$

```{figure} ../images/relat_12.png
:alt: relat_12.png
:width: 400px
SRI de la nave.
```

Por lo tanto, si vamos suficientemente rápido, podríamos llegar a cualquier punto del universo antes de morir de viejos, aunque estuvieran a millones de años-luz de distancia. Por supuesto, el tiempo en la tierra seguiría avanzando. Este efecto se ha comprobado experimentalmente, verificando que el tiempo de semidesintegración de las partículas es mayor si se desplazan más rápido.


## Paradoja del granjero y su hija

* Un granjero tiene una escalera que, después de entrar en el granero en el que quiere guardarla, sobresale un poco y no permite cerrar la puerta. Con el objetivo de poder cerrar la puerta, le pide a su hija que se aproxime al granero a una velocidad relativista $v$, para que desde el Sistema de Referencia Inercial (SRI) del granero en el que él se encuentra, la escalera se acorte por el efecto relativista de contracción espacial, y así él pueda cerrar la puerta.

$$
\Delta x = \frac{\Delta x'}{\gamma}
$$


```{figure} ../images/relat_21.png
:alt: relat_21.png
:width: 600px
SRI del granero.
```

Sin embargo, la hija del granjero le responde que, desde su propio Sistema de Referencia Inercial después de alcanzar la velocidad relativista $v$ necesaria, es el granero el que se aproxima a ella, por lo que el granero es el que se acorta por efecto relativista de contracción espacial. De esta forma, la escalera no puede entrar en el granero.


```{figure} ../images/relat_22.png
:alt: relat_22.png
:width: 400px
SRI de la escalera.
```

**Ambos tienen razón**. La aparente paradoja ocurre por la rotura de la simultaneidad. Podemos distinguir dos eventos: A) el contacto del extremo derecho de la escalera con el fondo del granero, y B) el paso del principio de la escalera por debajo del umbral de la puerta del granero. Si la escalera entra en el granero a una velocidad no relativista, ambos suceso ocurren prácticamente a la vez. Si pusiéramos dos relojes sincronizados tanto al principio como al final de la escalera, y otro par en la pared y en la puerta del granero, veréamos la rotura de la sincronía:

1) Si la escalera se aproxima a velocidad relativista, el granjero desde el SRI del granero ve que el evento B ocurre antes que el A. El granjero verá los relojes del granero sincronizados, pero verá los relojes de la escalera desincronizados.

2) Para la hija, desde el SRI de la escalera, el evento A ocurre antes que el B. Los relojes de la escalera estarán sincronizados pero los relojes del granero estarán desincronizados.

```{figure} ../images/relat_23.png
:alt: relat_23.png
:width: 400px
SRIs y simultaneidad.
```

El resultado final debemos dejarlo a la imaginación, ya que la colisión ocurrirá seguro, y sus consecuencias dependerán de la resistencia a la rotura y de la flexibilidad de tanto la pared del granero como de la escalera.


## Paradoja de los gemelos

* Un viajero espacial tiene una hermana melliza cuando emprende un viaje a velocidad relativista. Después de haber recorrido una cierta distancia vuelve a la tierra y se reencuentra con su hermana. El viajero ha envejecido menos que su hermana. Mientras que para él ha pasado un año, en la tierra han pasado dos. Esto se explica por el efecto de dilatación temporal. El tiempo que desde el Sistema de Referencia de la nave es un año, desde el Sistema de Referencia Inercial (SRI) de la tierra son 2 años.


```{figure} ../images/relat_31.png
:alt: relat_31.png
:width: 400px
SRI del granero.
```

Cabe preguntarse: Desde el Sistema de Referencia de la nave es la tierra la que se aleja y se acerca. ¿Por qué no es el tiempo de la tierra el que pasa más lento? Aquí ocurre una sutileza, y es que la nave ha tenido que dar la vuelta, por lo que en algún punto de su trayectoria, ha tenido que sufrir una aceleración y por lo tanto convertirse en un Sistema de Referencia No Inercial (SRNI).

1) Durante la primera parte del trayecto, mientras se aleja a velocidad constante, desde el SRI de la tierra, el tiempo en la nave avanza más lento, mientras que desde el SRI de la nave, el tiempo de la tierra avanza más lento.

2) En la maniobra acelerada para dar la vuelta, el tiempo en la nave transcurre de una manera que no se puede explicar por la teoría de la relatividad especial, que sólo trata con SRI. **El viajero en la nave sabe que está en un SRNI mientras realiza la maniobra, porque percibe fuerzas no inerciales**, mientras que la hermana en la tierra sabe que está en un SRI porque no es capaz de medir fuerzas no inerciales.

3) Durante el retorno a la tierra, vuelve a ocurrir lo mismo que el la primera parte, desde el SRI de la tierra, el tiempo en la nave avanza más lento, mientras que desde el SRI de la nave, el tiempo de la tierra avanza más lento.

Es la aceleración la que rompe la simetría de las trayectorias y es el hermano viajero el que tiene un tiempo que ha avanzado más lento que el de su hermana en la tierra. Este efecto se ha comprobado experimentalmente, poniendo un reloj perfectamente sincronizado con otro y colocando uno de los dos relojes en una nave que va a dar vueltas alrededor de la tierra (*movimiento acelerado*). Se ha comprobado que el reloj acelerado tuvo un retraso medible con respecto al reloj de la tierra.
