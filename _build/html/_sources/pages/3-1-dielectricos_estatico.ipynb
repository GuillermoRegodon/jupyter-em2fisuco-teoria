# Materiales dieléctricos en régimen estático

Hasta ahora, desde el punto de vista electrostático, tenemos varias posibilidades para modelar una región del espacio en el que podemos definir una campo eléctrico $\vec{E}$:

* **Vacío**: Caracterizado por las ecuaciones de la electrostática y por la permitividad dieléctrica del vacío $\varepsilon_0$.

* **Materiales conductores eléctricos perfectos CEP)**: Caracterizados por la presencia de cargas con movilidad infinita (*con respecto a la escala de tiempos de los problemas que podemos estudiar*). Nos permiten imponer unas condiciones de contorno sobre las ecuaciones de la electrostática para resolver algunos problemas de forma más sencilla.

* **Materiales conductores**: Caracterizados por la presencia de unas cargas móviles pero que no se pueden mover tan rápido. Nos permiten utilizar la relación entre el campo eléctrico $\vec{E}$ y la densidad de corriente $\vec{J}$. Tienen implicaciones en la disipación de potencia y en la atenuación de las ondas electromagnéticas.

* **Materiales dieléctricos**: Caracterizados por cargas que se pueden desplazar un poco con respecto a una posición de equilibrio a la que están ligadas. *Se modelan utilizando la permitividad eléctrica del material o su permitividad relativa y utilizando la definición de vector desplazamiento $\vec{D}$*. Estas definiciones nos permiten modificar las leyes de la electrostática para no tener que escribir explícitamente las cargas ligadas, que quedan modeladas mediante la permitividad relativa del material. **Observemos que aún no hemos relacionado la dinámica microscópica de las cargas ligadas con el efecto macroscópico**: Este es el objetivo de este tema.

```{note}
El objetivo de este tema es relacionar la respuesta macroscópica de un material dieléctrico con la respuesta microscópica de las cargas ligadas.
```

También vamos a realizar varios **modelos teóricos** para ciertos conceptos, **y vamos a extraer algunas conclusiones a partir de ellos**.

## Polarización

Vamos a definir la polarización $\vec{P}$ de la siguiente manera: *un dipolo es una configuración microscópica en la que tenemos una carga positiva $+q$ y una negativa $-q$ separadas una distancia $d$, de tal forma que el dipolo tiene un momento dipolar $\vec{p}$*:

$$
\vec{p} &= q(\vec{r_{+q}} - \vec{r_{-q}})\\
p &= qd
$$

**La polarización $\vec{P}$ se obtiene sumando todos los dipolos en una región de volumen $\Delta V$**:

$$
\vec{P} = \lim_{\Delta V \rightarrow 0} \frac{\sum \vec{p_i}}{\Delta V} = n <\vec{p}>
$$

Observemos que las unidades de la polarización son $\frac{C}{m^2}$. Esta definición es la más conveniente, porque como también vimos, las polarización de un material se puede sustituir por unas densidades de carga de polarización $\rho_p$ y $\sigma_p$, y resolver los problemas electrostáticas utilizando dichas densidades de carga. Siendo $\vec{n}$ el vector saliente de la superficie que envuelve un volumen con polarización $\vec{P}$:

$$
\rho_p &= -\nabla \cdot \vec{P} \\
\sigma_p &= \vec{P} \cdot \vec{n}
$$

**En la mayoría de los materiales, la polarización es proporcional al campo eléctrico $\vec{E}$**, de tal forma que tenemos la siguiente expresión:

$$
\vec{P} = \chi_e \vec{E}
$$

Finalmente, escribimos la definición de vector desplazamiento $\vec{D}$ y el resto de expresiones para la permitividad eléctrica del material $\varepsilon$ y la permitividad relativa del material $\varepsilon_r$.

$$
\vec{D} = \vec{P} + \varepsilon_o \vec{E} = \varepsilon_0 (1 + \chi_e) \vec{E} = \varepsilon_o \varepsilon_r \vec{E} = \varepsilon \vec{E}
$$

```{note}
En todo este recordatorio, la única relación entre lo microscópico y lo macroscópico es la que da la expresión:

$$
\vec{P} =  n <\vec{p}>
$$
```

## Polarizabilidad

Supongamos que tenemos un átomo neutro en una región de volumen. Tendrá cargas positivas y negativas $+q$ y $-q$ que se podrán desplazar una cierta distancia $d$ cuando apliquemos un campo eléctrico $\vec{E_L}$, generando un momento dipolar. Si, **como se comprueba experimentalmente para la mayoría de materiales, la relación entre $\vec{E}$ y $\vec{P}$ es proporcional**, podemos entender que la relación entre $\vec{p}$ y $\vec{E_L}$ también lo será: definimos así la polarizabilidad $\alpha$:

$$
\vec{p} = \alpha \vec{E_L}
$$

En lugar de $\vec{E}$, hemos utilizado $\vec{E_L}$ al que llamamos **campo eléctrico local**, mientras que conservamos $\vec{E}$ para el campo eléctrico impuesto desde el exterior. **No tienen por que ser iguales, $\vec{E} \neq \vec{E_L}$ , ya que el efecto de las cargas ligadas en general modifica el campo eléctrico total que actúa en la región del átomo**.

```{note}
**Modelo de gas diluido**

**Si podemos afirmar que los dipolos no se afectan mútuamente**, podemos decir que en este caso, el campo eléctrico local es igual al campo eléctrico externo:

$$
\vec{E} = \vec{E_L}
$$

Con esta premisa, podemos calcular la relación entre la polarización $\vec{P}$ y el campo eléctrico $\vec{E}$ para un gas a presión suficientemente baja.

$$
\vec{p} = \alpha \vec{E_L} \approx \alpha \vec{E} \quad \Rightarrow \quad \vec{P} = n \vec{p} = n \alpha \vec{E}
$$

Con lo que podemos deducir la relación entre el vector desplazamiento $\vec{D}$ y el campo eléctrico $\vec{E}$ y por tanto **la permitividad eléctrica de una gas a baja presión**.

$$
\vec{D} = \vec{P} + \varepsilon_0 \vec{E} = \varepsilon_0 (1 + n \alpha) \vec{E} = \varepsilon \vec{E}
$$

O lo que es lo mismo, que la permitividad eléctrica relativa $\varepsilon_r = 1 + \frac{n \alpha}{\varepsilon_0}$ y que $\chi_e = \frac{n \alpha}{\varepsilon_0}$.
```

Existen tres posibles motivos para que los átomos y moléculas se puedan polarizar, y por lo tanto tres contribuciones posibles a la hora de calcular la polarizabilidad $\alpha$.


### Polarizabilidad por distorsión eléctronica

En este mecanismo de polarización, la nube electrónica se desplaza un poco con respecto al núcleo positivo, de forma que se crea un dipolo. La distancia que se desplaza está relacionada con el campo eléctrico local $\vec{E_L}$. Para calcular la polarizabilidad por distorsión electrónica $\alpha_e$ vamos a realizar un pequeño modelo en el que consideramos la nube electrónica con una densidad de carga constante y buscamos el punto de equilibrio tras aplicar el campo eléctrico local $\vec{E_L}$.

````{note}
**Modelo de nube electronica con densidad de carga constante**

Supongamos que tenemos un átomo con $Z$ protones y $Z$ electrones, y supongamos que la nube electrónica tiene una densidad de carga constante de valor $rho$, con $R$ el radio del átomo:

$$
\rho = \frac{Q}{V} = \frac{-Ze}{\frac{4}{3} \pi R^3}
$$

El campo eléctrico generado por una densidad de carga constante es un problema que podemos resolver mediante Gauss, como se hizo en EMI. El módulo del campo eléctrico aumenta cuando nos alejamos del centro. Por esto, la posición de reposo es con el núcleo de carga $Ze$ en el centro.

$$
E_r = \frac{-Ze r}{4 \pi \varepsilon_0 R^3}
$$

Ahora supongamos que imponemos una campo eléctrico adicional $\vec{E_L}$. El núcleo se desplazara a la posición $r=d$ que cumple la condición de que la fuerza neta es nula, o lo que es lo mismo, el campo eléctrico generado por la nube electrónica $E_r$ se compensa con el campo eléctrico local $\vec{E_L}$.

```{figure} ../images/mat_1.png
:alt: mat_1.png
:width: 250px
Modelo de nube electrónica con densidad de carga constante.
```

$$
E_L = \frac{-Ze d}{4 \pi \varepsilon_0 R^3}
$$

El momento dipolar del dipolo que se forma es por tanto:

$$
p = qd = Zed = 4 \pi \varepsilon_0 R^3 E_L
$$

De donde deducimos que $\alpha_e = 4 \pi \varepsilon_0 R^3$
````

Por supuesto, este modelo se puede mejorar si calculamos la densidad electrónica real mediante los métodos de la física cuántica y la utilizamos en el modelo, pero la metodología es la misma.



### Polarizabilidad por distorsión iónica

Las moléculas están constituidas por átomos colocados en una posición de equilibrio, y, en la mayoría de los casos, los distintos átomos van a tener distinta carga. De esta forma, las moléculas pueden deformarse ligeramente con respecto a la posición de equilibrio. Este es el mecanismo de polarización por distorsión iónica $\alpha_i$. Para poder calcularla, necesitamos estimar la fuerza con la que la posición de reposo se mantiene y calcular cuál es la nueva posición de equilibrio al aplicar el campo eléctrico local $\vec{E_L}$.

```{figure} ../images/mat_2.png
:alt: mat_2.png
:width: 350px
Polarizabilidad por distorsión iónica.
```

Por lo tanto, para moléculas con iones de carga $+q$ y $-q$ aparece un dipolo adicional de valor $p = qd$. Obsérvese que el dipolo formado es el mismo independientemente de la orientación de la molécula, por lo que una sustancia puede tener una polarización propia neta nula, y polarizarse por distorsión iónica, siempre que esté constituido por moléculas con iones de cargas no nulas.

```{figure} ../images/mat_3.png
:alt: mat_3.png
:width: 250px
Polarizabilidad por distorsión iónica.
```

Experimentalmente, este mecanismo es difícil de distinguir del mecanismo de polarización por distorsión electrónica, y de hecho, ambos suelen agruparse en un único mecanismo de polarización por distorsión $\alpha_d = \alpha_e + \alpha_i$.



### Polarizabilidad por orientación

Cuando las moléculas que forman una sustancia tienen un momento dipolar $\vec{p_0}$ no nulo y la posibilidad de girar, van a tener tendencia a orientarse en la dirección del campo eléctrico local $\vec{E_L}$. En este caso, **la polarización máxima que pueden tener es precisamente la de todos los dipolos orientados**:

$$
\vec{P_{max}} = n \vec{p_0}
$$

En este caso, la energía de todos los dipolos es mínima. Sin embargo, si la sustancia tiene una cierta temperatura, los dipolos intentarán orientarse debido al efecto del campo eléctrico local $\vec{E_L}$, pero las colisiones y vibraciones aleatorias producirán una orientación aleatoria. **Los principios de la termodinámica nos dicen que el estado de equilibrio viene dado por la distribución de Maxwell-Boltzmann**. Para ello, necesitamos la energía de cada dipolo con orientación arbitraria, $\vec{p}$ con $p = p_0$  :

$$
U_p = -\vec{p} \cdot \vec{E_L}
$$

De esta forma, el número de partículas con una orientación determinada es proporcional a:

$$
N(\vec{p}) \propto \exp \left(-\frac{U_p}{k_BT} \right)
$$

Este problema tiene como solución la conocida como función de Langevin $\mathcal{L}(y) = \coth{y} - \frac{1}{y}$ , con $\vec{p_0}$ el momento dipolar de cada molécula, orientado en la dirección del campo eléctrico local $\vec{E_L}$:

$$
<\vec{p}> = \vec{p_0} \mathcal{L}\left( \frac{p_0 E_L}{k_BT} \right)
$$

En el límite para bajas temperaturas, la función de Langevin tiende a 1, por lo que efectivamente, predice que todos los dipolos se orientan en la dirección del campo eléctrico local $\vec{E_L}$. **Para temperaturas suficientemente altas**, la podemos aproximar por $\mathcal{L}(y) = \coth{y} - \frac{1}{y} \approx \frac{y}{3}$. De hecho, **en la mayoría de las situaciones, podremos utilizar esta aproximación**.


```{figure} ../images/mat_4.png
:alt: mat_4.png
:width: 400px
Función de Langevin.
```

$$
<p> &\approx p_0 \left( \frac{p_0 E_L}{3 k_BT} \right) =  \frac{p_0^2 E_L}{3 k_BT} \\
\alpha_o &= \frac{p_0^2}{3 k_BT}
$$

```{note}
Obsérvese que para poder realizar este cálculo, tan solo hemos necesitado la energía de cada orientación y la distribución de Maxwell-Boltzmann.
```



### Polarizabilidad total

La polarizabilidad total es la suma de las tres componentes:

$$
\alpha_T = \alpha_e + \alpha_i + \alpha_o = \alpha_d + \alpha_o
$$

Hemos visto que la polarizabilidad por distorsión, no depende de la temperatura, mientras que la polarizabilidad por orientación solo aparece si $p_0$ es distinto de cero y si depende de la temperatura.

$$
\alpha_T = \alpha_d + \frac{p_0^2}{3 k_BT}
$$

Podemos sacar varias conclusiones:

* La polarización por orientación, solo se puede dar en sustancias polares ($p_0 \neq 0$)

* La polarización por orientación aumenta con el inverso de la temperatura, por lo que para una temperatura suficientemente baja, siempre va a ser mayor que la polarización por distorsión (*En realidad, la función de Langevin dejaría de poder aproximarse por $\mathcal{L}(y) \approx \frac{y}{3}$, pero en todas las sustancias polares, dicha aproximación sigue siendo válida cuando $\alpha_o$ supera a $\alpha_d$*)

* Por regresión lineal, experimentalmente podemos distinguir $\alpha_o$ de $\alpha_d$, pero no podemos distinguir $\alpha_e$ de $\alpha_i$

```{figure} ../images/mat_5.png
:alt: mat_5.png
:width: 400px
Ejemplo de medidas experimentales de polarizabilidad.
```


## Materiales ferroeléctricos

Hasta ahora no hemos hecho el esfuerzo de calcular el campo eléctrico local $\vec{E_L}$ en función del tipo de material, sino que tan solo lo hemos aproximado por el campo eléctrico externo $\vec{E}$ en el caso de gas diluido. Esta aproximación es válida en ocasiones, **pero no explica fenómenos experimentales como son los materiales ferroeléctricos** que conservan polarización distinta de 0 aunque el campo eléctrico aplicado sea nulo.

**Para poder explicar la existencia de los materiales ferroeléctricos necesitamos un modelo que tenga en cuenda el efecto de los demás dipolos del material**, de forma que unos dipolos realimentan la polarización de sus dipolos vecinos y pueda haber polarización no nula.

El modelo más sencillo que tiene en cuenta el efecto de los dipolos cercanos es el modelo de Lorentz:

````{note}
**Modelo de Lorentz**

En este modelo, el material se modela como un conjunto de dipolos orientados en la misma dirección. Para los dipolos cercanos se calcula el campo eléctrico real que generan, mientras que para los dipolos lejanos, se utilizan las densidades de carga de polarización. Se define una frontera esférica de radio arbitrario que incluya suficientes dipolos cercanos, llamada la esfera de Lorentz.

```{figure} ../images/mat_6.png
:alt: mat_6.png
:width: 300px
Modelo de Lorentz.
```

De esta forma el campo eléctrico local $\vec{E_L}$ será **la suma de tres componentes**: El campo eléctrico externo $\vec{E}$, el campo eléctrico generado por los dipolos cercanos $\vec{E'}$ y el campo eléctrico generado por las densidades de carga de polarización $\vec{E_S}$.

$$
\vec{E_L} = \vec{E} + \vec{E'} + \vec{E_S}
$$

**Respecto a $\vec{E_S}$**, generado por las densidades de carga de polarización, éstas se pueden obtener utilizando las expresiones apropiadas, utilizando una polarización $\vec{P}$ constante y en la misma dirección que el campo eléctrico externo:

$$
\rho_p &= -\nabla \cdot \vec{P} = 0 \\
\sigma_p &= \vec{P} \cdot \vec{n} = P \vec{u_z} \cdot (-\vec{u_r}) = -P \cos \theta
$$

```{figure} ../images/mat_7.png
:alt: mat_7.png
:width: 300px
Modelo de Lorentz.
```

El campo generado por esta distribución de carga superficial en el centro es un problema que se resuelve utilizando las técnicas de EMI, y da como resultado:

$$
\vec{E_S} = \frac{\vec{P}}{3\varepsilon_0}
$$

**Respecto a $\vec{E'}$, el campo eléctrico de los dipolos cercanos, es nulo**. Podemos ver que por la ley de Gauss, la integral de superficie esférica del campo eléctrico generado por un dipolo debe ser 0.

$$
\varepsilon_0 \int_S \vec{E_p} \cdot \vec{dS} = Q_{enc} = 0
$$

Ahora podemos ver que la integral del campo eléctrico generado por una esfera de dipolos en el centro de dicha esfera, es exactamente la misma integral, intercambiando la posición del dipolo y del punto donde se realiza la suma. Para cada esfera de dipolos da 0, por lo que para toda la esfera sólida de dipolos también da cero.

Por lo tanto, uniendo los resultados:

$$
\vec{E_L} = \vec{E} + \frac{\vec{P}}{3\varepsilon_0}
$$

Por lo que la relación entre $\vec{P}$ y $\vec{E}$, usando la definición de $\chi_e$ (es decir, $\vec{P} = \varepsilon_0 \chi_e \vec{E}$) y de la polarizabilidad $\alpha$ (es decir, $\vec{P} = n \alpha \vec{E_L}$):

$$
\vec{E} = \frac{\vec{P}}{3\varepsilon_0} - \vec{E_L} = \vec{P} \left( \frac{1}{3\varepsilon_0} - \frac{1}{n\alpha} \right) = \varepsilon_0 \chi_e \vec{E} \left( \frac{1}{3\varepsilon_0} - \frac{1}{n\alpha} \right)
$$

Por lo que finalmente:

$$
\chi_e = \frac{\frac{n\alpha}{\varepsilon_0}}{1- \frac{n\alpha}{3\varepsilon_0}}
$$
````

Vemos que en el denominador hay una resta, lo que permite que la expresión de $\chi_e$ diverja. Es decir, el posible que exista una polarización no nula, para campo eléctrico externo nulo.

$$
\vec{P} = \varepsilon_0 \chi_e \vec{E} = \infty \cdot 0
$$

Utilizamos la polarizabilidad por orientación, $\alpha = \alpha_o = \frac{p_0^2}{3 k_BT}$ que depende de la temperatura, y de esta forma deducimos la conocida como Ley de Curie-Weiss, con $T_c = \frac{np_0^2}{9\varepsilon_0k_B}$:

$$
\chi_e = \frac{3 Tc}{T- Tc}
$$

De esta forma, utilizando el modelo de Lorentz y la polarizabilidad por orientación, se predice que para temperaturas inferiores a una temperatura crítica $T_c$, el material puede mantener la polarización $\vec{P}$ aunque el campo externo $\vec{E}$ sea nulo.
