# Potenciales electromagnéticos

**Los potenciales son una herramienta conveniente para resolver problemas electrostáticos**, y además, como se verá en la asignatura Electrodinámica Clásica, son *la manera natural de resolver problemas de fuentes (cargas $\rho$ y corrientes $\vec{J}$) que varían en el tiempo* cumpliendo las leyes de la relatividad especial.


## Potencial escalar en electrostática

El campo eléctrico $\vec{E}$ es irrotacional (en electrostática), es decir, $\nabla \times \vec{E} = 0$. Por lo tanto, sabemos que **el campo eléctrico $\vec{E}$ es conservativo y que es el gradiente de algún campo escalar**. El potencial escalar $\varphi$ surge como el campo escalar a partir del que se obtiene el campo eléctrico.
* Se define a través del campo eléctrico: $\vec{E} = -\nabla \varphi$.
* Para una carga puntual se puede calcular, siendo $r$ la distancia hasta la posición de la carga puntual:

$$
\varphi = \frac{q}{4 \pi \varepsilon_0 r}
$$

* Por superposición, para cualquier distribución:

$$
\varphi = \int\frac{\rho(\vec{r'})}{4 \pi \varepsilon_0 |\vec{r} - \vec{r'}|}
$$

* Dado que se define a partir del campo electrico a través de su derivada, no es único, ya que sumar una constante permite obtener el mismo campo eléctrico $\vec{E}$.

Aplicando esta definición a la ley de Gauss en el caso de electrostática:

$$
\nabla \cdot \vec{E} = - \nabla^2 \varphi = \frac{\rho}{\varepsilon_0}
$$

Por lo que tenemos **una ecuación que se puede plantear para obtener el potencial $\varphi$**. Es una ecuación muy estudiada, *que aparece en muchas ramas de las matemáticas y la física*. Tiene la gran ventaja de que *solo tiene un grado de libertad para cada punto del espacio*, mientras que la ley de Gauss para el campo eléctrico tiene 3 grados de libertad para cada punto del espacio, *haciendo que sea más facil de resolver*.

```{note}
**En electrostática**:

$$
\nabla \cdot \vec{E} = - \nabla^2 \varphi = \frac{\rho}{\varepsilon_0}
$$

```

## Potencial vector en magnetostática

El campo magnético $\vec{B}$ es solenoidal, es decir, tiene divergencia nula, $\nabla \cdot \vec{B} = 0$. Por lo tanto, **el campo magnético $\vec{B}$ es el rotacional de algún otro campo vectorial**, ya que la divergencia del rotacional siempre es nula, $\nabla \cdot (\nabla \times \vec{A}) = 0$, para cualquier campo vectorial $\vec{A}$. Definimos el potencial escalar utilizando esta propiedad a partir del campo magnético:
* El potencial vector se define a partir del campo magnético: $\vec{B} = \nabla \times \vec{A}$.
* A partir de la ley de Biot-Savart, podemos obtener el potencial vector de un hilo de corriente $I$ que recorre el camino $C$:

$$
\vec{A} = \frac{\mu_0 I}{4 \pi} \int \frac{\vec{dl'}}{|\vec{r} - \vec{r'}|}
$$

* Tiene definiciones parecidas para densidades de corriente $\vec{J}$ y densidades superficiales de corriente $\vec{j}$.

$$
\vec{A} = \frac{\mu_0}{4 \pi} \int \frac{\vec{J}(\vec{r'}) dV'}{|\vec{r} - \vec{r'}|}
$$

* Dado que se define a partir de su rotacional, tampoco es único, ya que sumar una constante permite obtener el mismo campo magnético $\vec{B}$.

**Tiene la ventaja de que la integral tiene una geometría más sencilla que la ley de Biot-Savart**. De hecho, la integral para el potencial vector es exactamente igual que la integral del potencial escalar, para cada componente de la densidad de corriente. Dicho de otra forma, **si descomponemos las componentes $x$, $y$ y $z$ de la ecuación integral para $\vec{A}$, obtenemos la misma ecuación que para $\varphi$** (salvo constantes multiplicativas).

```{note}
La ley de Biot-Savart tiene un producto vectorial dentro de la integral, por lo que solo se puede plantear de forma sencilla en casos con simetrías ventajosas.
```


## Potencial escalar y potencial vector generales

En el caso de las ecuaciones de Maxwell completas, **el campo magnético sigue siendo solenoidal, por lo que la definición de $\vec{A}$ no tiene por qué cambiar**:

$$
\vec{B} = \nabla \times \vec{A}
$$

Sin embargo, el rotacional del campo eléctrico no es nulo siempre:

$$
\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}
$$

Si usamos la definición del potencial vector, vemos que podemos encontrar otro campo modificado que si es irrotacional:

$$
\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t} = -\frac{\partial \nabla \times \vec{A}}{\partial t} \\
\nabla \times \left( \vec{E} + \frac{\partial \vec{A}}{\partial t} \right) = 0
$$

Este campo modificado, que se convierte en el campo eléctrico $\vec{E}$ en el caso de electrostática, es el que identificamos como el gradiente de un campo escalar $\vec{E} + \frac{\partial \vec{A}}{\partial t} = -\nabla \varphi$. Con esto:

$$
\vec{E} = -\nabla \varphi - \frac{\partial \vec{A}}{\partial t}
$$

```{note}
**Las expresiones generales de los campos electromagnéticos en función de los potenciales son**:

$$
\vec{B} &= \nabla \times \vec{A} \\
\vec{E} &= -\nabla \varphi - \frac{\partial \vec{A}}{\partial t}
$$

```

## Nuevas ecuaciones para los potenciales

Hasta ahora, no hemos planteado una ecuación que deba verificar el potencial vector $\vec{A}$, y siempre lo hemos calculado a partir de la expresión integral. Sin embargo, para el potencial escalar, si hemos planteado la ecuación de Poisson (muy estudiada, ya que aparece en muchas ramas de las matemáticas y la física), para electrostática:

$$
\nabla \cdot \vec{E} = -\nabla^2 \varphi = \frac{\rho}{\varepsilon_0}
$$

Esta ecuación se deduce facilmente a partir de la ley de Gauss. **Pero en electromagnetismo, la ecuación es incompleta**. La completamos:

$$
\nabla \cdot \vec{E} = -\nabla \cdot \left( -\varphi - \frac{\partial \vec{A}}{\partial t} \right) = \frac{\rho}{\varepsilon_0} \\
\nabla^2 \varphi + \frac{\partial }{\partial t} \nabla \cdot \vec{A}= - \frac{\rho}{\varepsilon_0}
$$

Sustituyendo las definiciones de los potenciales escalar y vector en la ley de Ampere generalizada, obtenemos la otra ecuación que cumplen los potenciales:

$$
\nabla \times \vec{B} = \mu_0 \vec{J} + \mu_0 \varepsilon_0 \frac{\partial \vec{E}}{\partial t} \\
\nabla \times \left( \nabla \times \vec{A} \right) = \mu_0 \vec{J} + \mu_0 \varepsilon_0 \frac{\partial}{\partial t} -\varphi - \frac{\partial \vec{A}}{\partial t}
$$

O lo que es lo mismo:

$$
\nabla^2 \vec{A} - \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} - \nabla \left( \nabla \cdot \vec{A} + \mu_0 \varepsilon_0 \frac{\partial \varphi}{\partial t} \right) = -\mu_0 \vec{J}
$$

**Estas son las ecuaciones que relacionan las fuentes con los potenciales**, pero un vistazo nos permite admitir que **¡no parecen muy fáciles de resolver!** Dado que los potenciales aparecen como derivadas, tienen esa libertad de sumar una constante y seguir ofreciendo los mismos campos electromagnéticos. De hecho ofrecen mayor libertad que esa, aunque no sea aparente. **Veamos que podemos usar esta libertad para simplificar las expresiones de una forma más práctica**.

```{note}
**En electromagnetismo general**:

$$
\nabla^2 \varphi &+ \frac{\partial }{\partial t} \nabla \cdot \vec{A}= - \frac{\rho}{\varepsilon_0} \\
\nabla^2 \vec{A} &- \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} - \nabla \left( \nabla \cdot \vec{A} + \mu_0 \varepsilon_0 \frac{\partial \varphi}{\partial t} \right) = -\mu_0 \vec{J}
$$

```
## Transformación general entre potenciales

Queremos relacionar dos pares de potenciales $\varphi$ y $\vec{A}$ con $\varphi'$ y $\vec{A'}$ de forma que ambos generen los mismos campos electromagnéticos:

$$
\vec{B} &= \nabla \times \vec{A} &=& \nabla \times \vec{A'}\\
\vec{E} &= -\nabla \varphi - \frac{\partial \vec{A} }{\partial t} &=&-\nabla \varphi' - \frac{\partial \vec{A'} }{\partial t}
$$

(*La transformación sencilla de sumar una constante, lo cumple $\varphi' = \varphi + C$ y $\vec{A'} = \vec{A} + \vec{C}$*)

Para obtener la transformación más general elegimos **un campo escalar cualquiera $C(\vec{r},t)$**, le sacamos el gradiente y se lo sumamos a $\vec{A}$. Como el rotacional de un gradiente es 0, la ecuación para $\vec{B}$ se cumple con $\vec{A}$ y con $\vec{A'}$.

$$
\vec{A'} = \vec{A} + \nabla C(\vec{r},t)
$$

Si introducimos esta igualdad en la ecuación para $\vec{E}$ obtenemos la igualdad que deben cumplir $\varphi$ y $\varphi'$.

$$
-\nabla \varphi - \frac{\partial \vec{A} }{\partial t} = -\nabla \varphi' - \frac{\partial}{\partial t} \left( \vec{A} + \nabla C(\vec{r},t) \right) \\
-\nabla \varphi = -\nabla \varphi' - \nabla \frac{\partial C(\vec{r},t)}{\partial t}
$$

Por lo que:

$$
\varphi' = \varphi - \frac{\partial C(\vec{r},t)}{\partial t}
$$

Estas expresiones nos permiten obtener un potencial a partir de otro, pero a parte de la elegancia y la belleza del resultado, ¡no nos resulta de utilidad! La estrategia que utilizamos en la práctica es:

* Imponemos alguna condición ventajosa para que las ecuaciones generales de los potenciales sean más fáciles de resolver.
* Comprobamos que existe un $C(\vec{r}, t)$ que permite que la simplificación sea factible.

**Observamos que lal ecuaciones generales de los potenciales son válidas para todos los pares de potenciales $\varphi$ y $\vec{A}$, mientras que las expresiones simplificadas solo vas a ser válidas para unos pocos**. Esta transformación es lo que llamamos *gauge* o *calibración*.


## Gauge de Coulomb

Si nos fijamos en las ecuaciones generales, vemos que la expresión de los potenciales en función de la carga se simplificaría a la ecuación de Poisson si pido que $\nabla \cdot \vec{A} = 0$. Por lo tanto:

$$
\nabla^2 \varphi + \frac{\partial }{\partial t} \nabla \cdot \vec{A}= - \frac{\rho}{\varepsilon_0}
$$

Pasaría a ser:

$$
\nabla^2 \varphi = - \frac{\rho}{\varepsilon_0}
$$

La pregunta es, ¿siempre existe un par de potenciales que cumple que $\nabla \cdot \vec{A} = 0$? La respuesta es si. De la transformación general:

$$
\vec{A'} = \vec{A} + \nabla C(\vec{r},t)
$$

Aplico divergencia e impongo que $\nabla \cdot \vec{A'} = 0$, por lo que:

$$
\nabla \cdot \vec{A} = - \nabla^2 C(\vec{r},t)
$$

El miembro de la izquierda es cualquier potencial válido, y el de la derecha es una ecuación de Poisson que siempre tiene solución, por lo que los potenciales $\varphi'$ y $\vec{A'}$ con $\nabla \cdot \vec{A'} = 0$ siempre se pueden obtener de un par de potenciales válidos $\varphi$ y $\vec{A}$.

**El gauge de Coulomb tiene la ventaja de que el potencial escalar $\varphi$ se resuelve igual que el electrostática**. La otra ecuación es un poco más fea, pero también es una ecuación escudiada y con solución.

$$
\nabla^2 \vec{A} - \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} = -\mu_0 \vec{J} + \nabla \left( \mu_0 \varepsilon_0 \frac{\partial \varphi}{\partial t} \right)
$$

```{note}
**En el gauge de Coulomb**:

$$
\nabla^2 \varphi &= - \frac{\rho}{\varepsilon_0} \\
\nabla^2 \vec{A} &- \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} = -\mu_0 \vec{J} + \nabla \left( \mu_0 \varepsilon_0 \frac{\partial \varphi}{\partial t} \right)
$$

```

## Gauge de Lorentz

El gauge de Lorentz surge de observar que si el paréntesis en la ecuación de los potenciales en función de la corriente fuera nulo, se simplificaría a una ecuación de onda no homogénea. Si se cummple:

$$
\nabla \cdot \vec{A} + \mu_0 \varepsilon_0 \frac{\partial \varphi}{\partial t} = 0
$$

entonces...

$$
\nabla^2 \vec{A} - \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} = -\mu_0 \vec{J}
$$

De la misma forma que antes, si aplico la divergencia sobre la ecuación de la transformación, me queda una ecuación de Poisson que siempre tiene solución, por lo que al igual que antes, a partir de cualquier par de potenciales validos, se puede obtener un par de poteniales que cumpla el gauge de Lorentz. La otra ecuación para los potenciales **¡resulta ser otra ecuación de onda! ¡Además con la velocidad de la luz como velocidad de propagación!**.

$$
\nabla^2 \varphi - \mu_0 \varepsilon_0 \frac{\partial^2 \varphi}{\partial t^2} = -\frac{\rho}{\varepsilon_0}
$$

```{note}
**En el gauge de Lorentz**:

$$
\nabla^2 \varphi - \mu_0 \varepsilon_0 \frac{\partial^2 \varphi}{\partial t^2} &= -\frac{\rho}{\varepsilon_0} \\
\nabla^2 \vec{A} - \mu_0 \varepsilon_0 \frac{\partial^2 \vec{A}}{\partial t^2} &= -\mu_0 \vec{J}
$$

```
