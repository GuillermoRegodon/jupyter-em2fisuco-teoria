# Materiales dieléctricos en régimen dinámico

Al estudiar los materiales, hemos hecho algunas suposiciones acerca de los tiempos que tardan en estabilizarse, comparados con la escala de tiempos a la que se producen las variaciones de los campos impuestos externamente:

* En los conductores (*caracterizados por $\vec{J_c} = \sigma \vec{E}$*) las cargan se aceleran desde el reposo hasta la velocidad final de forma prácticamente instantánea.

* En los Conductores eléctricos perfectos (*CEP*) las cargas se reorganizan de forma prácticamente instantánea para que el campo eléctrico sea perpendicular a la superficie y el campo magnético sea tangencial.

* En los materiales con una determinada permitividad eléctrica, los átomos y moléculas se polarizan microscópicamente de forma prácticamente instantánea. Obsérvese que en todos los caso se ha supuesto una situación de equilibrio para poder calcular la polarizabilidad $\alpha$.

Sin embargo, a frecuencias suficientemente altas y con variaciones externas de os campos suficientemente rápidas, estas aproximaciones pueden dejar de ser válidas: Los conductores eléctricos perfectos pueden dejar de serlo, los conductores no se podrán caracterizar por una conductividad...

En el caso de los dieléctricos, **los distintos mecanismos de polarización tienen escalas de tiempos y de frecuencias distintos**:

* **La polarización por distorsión electrónica**, que solo implica un desplazamiento de los electrones, tiene la escala de tiempos $\tau_e$ más rápida.

* **La polarización por distorsión iónica** implica el desplazamiento relativo de los iones de una molécula, por lo que tiene una escala de tiempos más lenta $\tau_i > \tau_e$. Está relacionada con los estados vibracionales de la molécula.

* **La polarización por orientación** implica la rotación de la molécula para acomodarse a la orientación del campo eléctrico externo, por lo que su escala de tiempos $\tau_o$ es la más lenta.

$$
\tau_e &< \tau_i < \tau_o \\
f_e &> f_i > f_o
$$

Vemos que es posible que los cambios externos sean demasiado rápidos, o la frecuencia de la excitación externa demasiado alta, como para que algún mecanismo de polarización tenga tiempo de producirse. En este caso, **dicho fenómeno no contribuye a formar los dipolos microscópicos y la permitividad eléctrica del material responde como si dicho mecanismo no existiera**. En términos de la frecuencia, la polarizabilidad pasa por los siguientes valores:

* Para $f$ suficientemente baja, $\alpha_T = \alpha_e + \alpha_i + \alpha_o$.
* En el rango de microondas, $f_o \approx 10^{10}$ Hz, las moléculas empiezan a no tener tiempo a orientarse, y  para frecuencias superiores $f_o \ll f \ll f_i$, el valor de la polarizabilidad $\alpha_T \rightarrow \alpha_e + \alpha_i$.
* En el rango infrarrojo, $f_i \approx 10^{13}$ Hz, la masa de los iones impide que se puedan desplazar respecto a su posición de equilibrio, y para frecuencias superiores $f_i \ll f \ll f_e$, el valor de la polarizabilidad $\alpha_T \rightarrow \alpha_e$.
* En el rango visible ocurre la última transición, $f_e \approx 10^{15}$, y para frecuencias en el ultravioleta, no tiene sentido estudiar los materiales utilizando una permitividad eléctrica del material, ya que $\alpha_T \rightarrow 0$.

Para estudiar el rango de frecuencias intermedio, vamos a utilizar **un modelo en el que distinguimos dos fenómenos de polarización, uno rápido $\tau_1$ y otro lento $\tau_2$**. Dadas las frecuencias de transición, solo tenemos dos posibilidades:

* En el rango de frecuencias de microondas, $\tau_1$ sería la polarización por distorsión (*electrónica más iónica*), y $\tau_2$ sería la polarización por orientación.
* En el rango de infrarrojos, $\tau_1$ sería la polarización por por distorsión electrónica, y $\tau_2$ sería la polarización por distorsión iónica.

````{note}
**Modelo de relajación dieléctrica, respuesta al escalón**

En este modelo, descomponemos la polarización en dos componentes, $\vec{P} = \vec{P_1} + \vec{P_2}$, siendo una de ellas suficientemente rápida para considerar que, en el rango de frecuencias del problema:

$$
\vec{P_1}(t) = \varepsilon_0 \chi_1 \vec{E}(t)
$$

La otra componente sigue un sistema de primer orden, en el que las soluciones son exponenciales decreciente:

$$
\frac{{\rm d} \vec{P_2}(t)}{{\rm d} t} = \frac{\varepsilon_0 \chi_2 \vec{E}(t) - \vec{P_2}(t)}{\tau_2}
$$

La solución ante un cambio brusco es una exponencial decreciente, como se puede verificar por sustitución. Para un escalón en el campo eléctrico, $\vec{E}(t) = \vec{E_0}$ para $t \ge 0$, y $\vec{E}(t) = 0$ para $t<0$.

$$
\vec{P_2}(t) = \varepsilon_0 \chi_2 \vec{E_0} \left(1 - \exp\left(-\frac{t}{\tau_2} \right) \right)
$$

La respuesta total en el tiempo, ante un escalón:

$$
\vec{P}(t) = \vec{P_1}(t) + \vec{P_2}(t) = \varepsilon_0 \left( \chi_1 + \chi_2 \left(1 - \exp\left(-\frac{t}{\tau_2} \right) \right) \right) \vec{E_0}
$$

````


````{note}
**Modelo de relajación dieléctrica, respuesta en frecuencia**

Vamos a resolver las misma ecuaciones, pero esta vez en el dominio de la frecuencia. La ecuación diferencial para $\vec{P_2}(t)$ se convierte en la siguiente ecuación de fasores:

$$
j \omega \tau_2 \vec{P_2} = \varepsilon_0 \chi_2 \vec{E} - \vec{P_2}
$$

Por lo tanto la relación entre $\vec{P}$ y $\vec{E}$ pasa a ser:

$$
\vec{P} = \vec{P_1} + \vec{P_2} = \varepsilon_0 \left( \chi_1 + \frac{\chi_2}{1 + j \omega \tau_2} \right) \vec{E}
$$

Vemos que, **cuando estudiamos fasores, tiene sentido hablar de una permitividad eléctrica del material compleja**:

$$
\varepsilon = \varepsilon_0 \left( 1 + \chi_1 + \frac{\chi_2}{1 + j \omega \tau_2} \right)
$$

Esta expresión compleja tiene dos casos límite reales:

* Para $\omega \ll 1/\tau_2$, la permitividad compleja es la de los dos mecanismos de polarización:

$$
\varepsilon(\omega \tau_2 \ll 1) = \varepsilon_0 \left( 1 + \chi_1 + \chi_2 \right)
$$

* Para $\omega \gg 1/\tau_2$, la permitividad compleja es la del mecanismo de polarización rápido:

$$
\varepsilon(\omega \tau_2 \gg 1) = \varepsilon_0 \left( 1 + \chi_1 \right)
$$

En el rango intermedio, la permitividad es compleja, $\varepsilon = \varepsilon' - j \varepsilon''$ (*Obsérvese el signo menos en la parte compleja, por definición*). Obviamente,

$$
\frac{\varepsilon'}{\varepsilon_0}  \in [\left( 1 + \chi_1 \right), \left( 1 + \chi_1 + \chi_2 \right)]
$$

La parte imaginaria se puede sacar de la expresión completa:

$$
\varepsilon = \varepsilon_0 \left( 1 + \chi_1 + \frac{\chi_2}{1 + j \omega \tau_2} \right) = \varepsilon_0 \left( 1 + \chi_1 + \frac{\chi_2(1 - j \omega \tau_2)}{1 + \omega^2 \tau_2^2} \right)
$$

Por lo que:

$$
\frac{\varepsilon''}{\varepsilon_0 } = \frac{\chi_2 \omega \tau_2)}{1 + \omega^2 \tau_2^2}
$$

La parte imaginaria de la permitividad eléctrica del material vale 0 en los casos límite antes mencionados, $\omega \tau_2 \ll 1$ y $\omega \tau_2 \gg 1$. En el rango intermedio tiene un máximo, que se puede obtener derivando $\frac{\varepsilon''}{\varepsilon_0}$ frente a $\omega \tau_2$ e igualando a 0. Después del cálculo, el máximo está en $\omega \tau_2 = 1$, y vale $\varepsilon'' = \varepsilon_0 \frac{\chi_2}{2}$. Para esta misma frecuencia, la parte real $\varepsilon' = \varepsilon_0 \left(1 + \chi_1 + \frac{\chi_2}{2} \right)$.

De hecho, si calculamos la distancia entre la permitividad y el punto complejo $\varepsilon_0 \left(1 + \chi_1 + \frac{\chi_2}{2} \right)$, da un valor constante, por lo que comprobamos que la permitividad eléctrica traza una circunferencia en el plano complejo.

$$
\left|\varepsilon - \varepsilon_0 \left(1 + \chi_1 + \frac{\chi_2}{2} \right) \right| &= \varepsilon_0 \chi_2 \left| \frac{1}{1 + j \omega \tau_2} -\frac{1}{2} \right| \\
&= \varepsilon_0 \chi_2 \left| \frac{1 - \frac{1}{2} (1 + j \omega \tau_2)}{1 + j \omega \tau_2}  \right| \\
&= \frac{\varepsilon_0 \chi_2}{2} \left| \frac{1 - j \omega \tau_2}{1 + j \omega \tau_2}  \right| \\
&= \frac{\varepsilon_0 \chi_2}{2} \frac{|1 - j \omega \tau_2|}{|1 + j \omega \tau_2|} =  \frac{\varepsilon_0 \chi_2}{2} 
$$


```{figure} ../images/mat_9.png
:alt: mat_9.png
:width: 500px
Respuesta en frecuencia del modelo de relajación dieléctrica.
```
```{figure} ../images/mat_10.png
:alt: mat_10.png
:width: 500px
Ejemplo de medidas experimentales usando el modelo de relajación dieléctrica.
```



¿Qué implicaciones tiene la permitividad compleja? En la ecuación de ondas, la parte compleja de la permitividad tiene la misma función que la conductividad, por lo que también implica unas perdidas, un desfase entre el campo eléctrico y el campo magnético... Vamos a verlo, por ejemplo, en la ecuación fasorial que planteamos para obtener las constantes de fase y de atenuación de una onda plana:

$$
(j \beta + \alpha)^2 + \mu \varepsilon \omega^2 - j \mu \sigma \omega = 0
$$

Vemos que si la permitividad eléctrica del material es una magnitud compleja, lo que hacemos es añadir un término imaginario puro a la conductividad:

$$
(j \beta + \alpha)^2 + \mu \varepsilon'\omega^2 - j \mu \left(\sigma + \varepsilon'' \omega \right) \omega = 0
$$

La conductividad y la parte imaginaria de la permitividad eléctrica del material están sumando. Para frecuencias bajas, $\omega \ll \frac{\sigma}{\varepsilon''}$, la conductividad $\sigma$ es el efecto predominante, pero para frecuencias altas, $\omega \gg \frac{\sigma}{\varepsilon''}$, las perdidas por la permitividad compleja superan a la conductividad.

````
