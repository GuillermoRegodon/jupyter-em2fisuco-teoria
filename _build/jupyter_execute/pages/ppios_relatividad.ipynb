# Principios de relatividad especial

Los postulados de Einstein, en torno a los que se construye la relatividad especial son los siguientes:

1. Las leyes físicas son las mismas en todos los sistemas de referencia inerciales (o SRI).
2. La velocidad de la luz $c$ es la misma desde todos los sistemas de referencia inerciales (o SRI).

Esto tiene como consecuencia varias cosas importantes:

1. **La dilatación temporal**: Si la diferencia de tiempo que pasa entre dos eventos es $\Delta t'$ desde un SRI en el que la posición donde ocurren los eventos es la misma, entonces la diferencia de tiempo $\Delta t$ que pasa entre los dos eventos visto desde un SRI en el que los eventos pasa en posiciones distintas es mayor en un factor $\gamma$.

$$
\Delta t = \gamma \Delta t'
$$

2. **La contracción espacial**: Si la longitud de un objeto que se mide desde un SRI en el que el objeto está en reposo es $\Delta x'$, la longitud del objeto $\Delta x$ visto desde un SRI en el que el objeto está en movimiento es menor en un factor $\gamma$.

$$
\Delta x = \frac{\Delta x'}{\gamma}
$$

Con $\gamma = \frac{1}{\sqrt{1-\beta^2}}$, con $\beta = \frac{v}{c}$ el módulo de la velocidad relativa entre los dos SRI en unidades de $c$.


```{note}
**El punto de vista de una partícula**

Los efectos relativistas de contracción espacial y dilatación temporal tienen una aplicación práctica importante cuando el sistema de referencia SRI' en movimiento está pegado a una partícula que se mueve a velocidad relativista respesto a nuestro sistema de referencia SRI en el laboratorio. **Si el tiempo de vida medio de una partícula es $t_{1/2}$, y la partícula se desplaza a velocidad $v$ medido en SRI, desde SRI el tiempo de vida medio es $\gamma t_{1/2}$**.

El efecto de contracción espacial dice que la partícula vería el laboratorio mas corto (*en la dirección de su desplazamiento*) en el mismo factor $\gamma$.
```

Estas transformaciones de distancias y tiempos se pueden poner de forma simétrica como una transformación de coordenadas, conocida como **Transformación de Lorentz**:

$$
ct' &= \gamma(ct - \beta x) \\
x' &= \gamma(x - \beta ct)
$$

La transformación de Lorentz era conocidad antes de los postulados de la relatividad y el trabajo de Einstein: La transformación de Lorentz se dedujo matemáticamente como la transformación necesaria para que la ley de inducción de Faraday fuera independiente del sistema de referencia. El hecho de que se deduzca a partir de los postulados de Einstein da justificación y validez a dichos postulados.

El último efecto relativista, que es de hecho el más fundamental, es **la rotura de la simultaneidad**. Para verlo, vamos a dibujar el Diagrama de Minkowski, que es simplemente un eje de coordenadas cartesianos en el que la ordenada es el tiempo (*multiplicado por c para que tenga unidades de distancia*). Es este diagrama, la trayectoria de un haz de luz a velocidad $c$ es una recta inclinada $45º$.

$$
ct = x
$$

```{figure} ../images/minkowski_1.png
:alt: minkowski_1.png
:width: 325px
Diagrama de Minkowski y haz de luz.
```

Los ejes $x',ct'$ del SRI' vistos desde el SRI están inclinados. Para obtener la expresión del eje $ct'$ en función de $x,ct$, impongo que $x'=0$.

$$
x' = 0 \Rightarrow ct = \frac{x}{\beta} 
$$

Igualmente, obtenemos la expresión del eje $x'$ en función de $x,ct$ imponiendo que $ct'=0$.

$$
ct' = 0 \Rightarrow ct = \beta x
$$

```{figure} ../images/minkowski_2.png
:alt: minkowski_2.png
:width: 325px
Diagrama de Minkowski y ejes de SRI'.
```

Ahora es fácil comprobar que **dos eventos que ocurren simultáneamente desde SRI, no ocurren simultáneamente desde SRI'**. Al revés también ocurre, que dos eventos que ocurren simultáneamente desde SRI', no ocurren simultáneamente desde SRI.


```{figure} ../images/minkowski_3.png
:alt: minkowski_3.png
:width: 625px
Diagrama de Minkowski y simultaneidad.
```

```{note}
Con la rotura de la simultaneidad, podemos explicar muchas aparentes paradojas de la relatividad:

* Cuando un objeto $A$ se mueve con velocidad relativista $v$ con respecto a otro $B$, el segundo, $B$, se mueve con velocidad relativista $-v$ con respecto al primero $A$. ¿Cómo es posible que $A$ vea más corto a $B$ y $B$ vea más corto a $A$? **Por que los instantes temporales en los que $A$ mira la parte de delante y de atrás de $B$ no coinciden con los instantes temporales en los que $B$ mira sus propios extremos, y viceversa**. Si pusieramos relojes en los extremos de los objetos $A$ y $B$, La sincronía entre los relojes depende de quién los mira. Claro, entonces $B$ podría pensar *"$A$ piensa que soy más corto de lo que realmente soy porque mide mal la distancia entre mis extremos"*. Pero por el primer postulado de Einstein: **Los dos puntos de vista son válidos y cumplen con todas las leyes de la física**.
```

La transoformación de Lorentz tiene un último resultado importante: En una transformación de Lorentz, se conserva la siguiente expresión, llamada intervalo:

$$
c^2t^2 - x^2 = c^2t'^2 - x'^2
$$

```{note}
**El punto de vista de una partícula**

El intervalo y la conservación del intervalo en una transformación relativista es muy importante porque el momento relativista $\vec{P}$ es un vector relativista de componentes $E/c = \gamma m c$ en la dirección del eje $ct$, y $p = \gamma m v$ en la dirección $x$ (*ver siguiente nota*). 


Por lo tanto, en una transformación de Lorentz, lo que se conserva es el intervalo del momento relativista:

$$
\frac{E^2}{c^2} - p^2 = \frac{E'^2}{c^2} - p'^2
$$

En particular, desde el SRI' pegado a la partícula, está en reposo (*$p'=0$, y su energía es la energía de reposo $E' = E_0 = mc^2$*), por lo que:

$$
\frac{E_0^2}{c^2} = m c
$$

Si realizamos la transformación relativista para ver la partícula desde nuestro SRI en el laboratorio, obtenemos la siguiente expresión:

$$
E^2 - p^2 c^2 = m^2 c^4
$$

**Esta es una de las expresiones más utilizadas en física de partículas relativistas, por que relaciona correctamente magnitudes que podemos medir desde nuestro SRI del laboratorio**.
```

```{note}
El momento relativista $\vec{P}$ asociado a una partícula se deduce de la definición de momento en función de la velocidad, y la definición de la velocidad en función del tiempo propio de la partícula, que es el tiempo desde su propio sistema de referencia:

$$
d\tau = dt' = \frac{dt}{\gamma}
$$

La velocidad relativista $\vec{U}$ es, por tanto:

$$
\vec{U} = \frac{(cdt, dx)}{d\tau} = \left( c\frac{dt}{d\tau}, \frac{dx}{d\tau} \right) = \left( c\frac{dt}{d\tau}, \frac{dx}{dt}\frac{dt}{d\tau} \right) = ( \gamma c, \gamma v)
$$

El momento relativista $\vec{P}$ se obtiene multiplicando por la masa de la partícula $\vec{P} = ( \gamma m c, \gamma m v)$. El uso de la energía $E/c$ en la coordenada temporal $\gamma m c$ es una idea que no se puede justificar a priori, pero que da uno de los resultados más conocidos de Einstein:

$$
\gamma m c = \frac{E}{c} \Rightarrow E_0 = E_{(v=0)} = mc^2
$$

Sin embargo, se justifica a posteriori, dado que si llamamos energía cinética a la energía adicional que posee la partícula debido a su movimiento, entonces obtenemos una expresión que se puede aproximar a la energía cinética clásica cuando las velocidades son pequeñas. De esta manera, la definición de la coordenada temporal $E/c$ del momento relativista está justificada.

$$
E_k &= E - E_0 = \gamma mc^2 - mc^2 \\
E_k &\approx \frac{1}{2} m v^2 \quad (v \ll c)
$$
```
