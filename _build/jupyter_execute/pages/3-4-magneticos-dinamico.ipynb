# Propiedades magnéticas de los medios materiales en régimen dinámico

Al estudiar los materiales, hemos supuesto que se alcanza un estado de equilibrio de forma suficientemente rápida con respecto a la escala de tiempos de la variación de los campos magnéticos externos. **En este tema, vamos a revisar las ecuaciones de la variación de los dipolos magnéticos** y vamos a resolverlas en algunos casos ilustrativos.

## Ecuación de Landau-Lifshitz

Un dipolo magnético, caracterizado por su momento dipolar magnético $\vec{m}$, inmerso en un campo magnético $\vec{B}$ sufre un torque $\vec{\tau}$ que tiende a orientarlo en la misma dirección de $\vec{B}$.

$$
\vec{\tau} = \vec{m} \times \vec{B}
$$

Por una parte, el torque $\vec{\tau}$ es precisamente la variación del momento angular $\frac{{\rm d}\vec{l}}{{\rm d}t}$, que a su vez está relacionado con el momento dipolar magnético $\vec{m} = -\frac{e}{2m_e} \vec{l} = -\gamma \vec{l}$. Por lo tanto:

$$
\frac{{\rm d} \vec{m}}{{\rm d} t} = - \gamma \vec{m} \times \vec{B}
$$

En un material, no todos los momentos dipolares magnéticos son de la misma magnitud, por lo que cada uno tendrá una dinámica un poco diferente, pero en primera aproximación podemos suponer que todos son iguales. Por lo tanto, para un volumen $\Delta V$ de material,

$$
\frac{{\rm d} \vec{M}}{{\rm d} t} = - \gamma \vec{M} \times \vec{B}
$$

Esta ecuación es conocida como la **ecuación de Landau-Lifshitz**. Vamos a resolverla en algunos casos.

```{note}
**Campo magnético constante $\vec{B} = B_0 \vec{u_z}$**

Estudiando las componentes de la ecuación:

$$
\frac{{\rm d} M_x}{{\rm d} t} &= - \gamma M_y B_0 \\
\frac{{\rm d} M_y}{{\rm d} t} &= \gamma M_x B_0 \\
\frac{{\rm d} M_z}{{\rm d} t} &= 0
$$

La primera conclusión es que $M_z = M_0$ es constante. La segunda se obtiene desacoplando las dos primeras ecuaciones.

$$
\frac{{\rm d^2} M_x}{{\rm d} t^2} &= - \gamma^2 B_0^2 M_x\\
\frac{{\rm d^2} M_y}{{\rm d} t^2} &= - \gamma^2 B_0^2 M_y
$$

Es decir, la imanación en la dirección transversal es una función armónica de frecuencia $\omega_0 = \gamma B_0$, conocida como la **frecuencia de Larmor**. Además, la ecuación original demuestra que el desfase entre las direcciones es $\pi/2$, por lo que la imanación transversal traza una circunferencia en torno al eje $z$.

$$
M_x &= M_1 \cos(\omega_0 t) \\
M_y &= M_1 \sin(\omega_0 t) \\
M_z &= M_0
$$
```

````{note}
**Campo magnético rotatorio $\vec{B} = B_0 \vec{u_z} + B_1 (\cos (\omega t) \vec{u_x} + \sin (\omega t) \vec{u_y})$**

En este caso, las ecuaciones por componentes:

$$
\frac{{\rm d} M_x}{{\rm d} t} &= - \gamma M_y B_0 + \gamma M_z B_1 \sin (\omega t)\\
\frac{{\rm d} M_y}{{\rm d} t} &= \gamma M_x B_0 - \gamma M_z B_1 \cos (\omega t)\\
\frac{{\rm d} M_z}{{\rm d} t} &= \gamma B_1 (M_x \sin (\omega t) - M_y \cos (\omega t))
$$

Aunque a priori, no tenemos un motivo para ello, vamos a realizar la suposición adicional de que $M_z = M_0$ constante.

```{note}
La suposición $M_z = M_0$ se puede justificar por los fenómenos de relajación que presenta el material y que hacen que la imanación acabe atenuándose, de forma que al final solo quedará la imanación inducida por el campo magnético rotatorio. Sin embargo, este efecto no se puede predecir a partir de la ecuación de Landau-Lifshitz, qie no tiene en cuenta fenómenos de relajación de la imanación.
```

Con esto, las otras dos ecuaciones se convierten en dos ecuaciones diferenciales de primer orden acopladas, que se pueden desacoplar y resolver:

$$
\vec{M} = M_0 \vec{u_z} + \frac{\gamma M_0}{\omega_0 - \omega}\vec{B_1}
$$

Donde, por simplicidad de la expresión, hemos usado $\vec{B_1} = B_1 (\cos (\omega t) \vec{u_x} + \sin (\omega t) \vec{u_y})$. Hemos obtenido el resultado de que la magnetización sigue al campo magnético en fase, pero solo puede seguirlo si existe el campo magnético longitudinal $B_0 \vec{u_z}$ que genera $M_0$. Si $M_0 = 0$, el material no puede imanarse con este campo.

Otro resultado interesante es que la imanación puede entrar en resonancia si la frecuencia de la excitación $\omega$ se acerca a la frecuencia de Larmor $\omega_0$.
````

## Ecuaciones de Bloch

En los materiales imanados, existen fenómenos de relajación que atenúan la imanación. En presencia de una campo magnético constante en la dirección $z$, encontramos dos tipos de relajación:

* **Relajación longitudinal**: Ocurre por la diferencia de energía entre los distintos niveles energéticos. Las vibraciones de la red cristalina o del resto del material tienen una probabilidad de intercambiar energía con estos niveles, con un tiempo característico $\tau_1$. El estado final sería el estado de equilibrio termodinámico característico de los materiales paramagnéticos, con una imanación final $M_0$. Se caracteriza como un sistema de primer orden con un valor final $M_0$:

$$
\frac{{\rm d} M_z}{{\rm d} t} = \frac{M_0 - M_z}{\tau_1}
$$

* **Relajación transversal**: Los materiales que se caracterizan por una imanación $\vec{M} = n \left<\vec{m}\right>$ están constituidos por distintos momentos dipolares magnéticos que en media valen $\left<\vec{m}\right>$. Dado que cada momento dipolar magnético tiene tendencia a precesionar a una frecuencia diferente, el desfase entre estos dipolos magnéticos aumenta. En un tiempo característico $\tau_2$ los dipolos acaban teniendo orientaciones opuestas y la media se anula, $\left<\vec{m}\right> \rightarrow 0$. Se caracteriza como un sistema de primer orden con valor final nulo

$$
\frac{{\rm d} M_{x,y}}{{\rm d} t} = -\frac{M_{x,y}}{\tau_2}
$$

El sistema global, incluyendo el torque, constituye las ecuaciones de Bloch:

$$
\frac{{\rm d} \vec{M}}{{\rm d} t} = - \gamma \vec{M} \times \vec{B} + \left( \begin{array}{c} -\frac{M_x}{\tau_2} \\ -\frac{M_y}{\tau_2} \\ \frac{M_0 - M_z}{\tau_1} \end{array} \right)
$$

````{note}
**Campo magnético constante $\vec{B} = B_0 \vec{u_z}$**

Cuando un material con una determinada imanación se relaja en presencia de una campo magnético constante $\vec{B} = B_0 \vec{u_z}$, el material se puede relajar de varias maneras, dependiendo de los valores de los tiempo de relajación, y como se relacionan con la frecuencia de Larmor, pero la solución más común es que se relaje hacia $\vec{M} \rightarrow M_0 \vec{u_z}$ dando vueltas alrededor del eje $z$.


```{figure} ../images/mat_16.png
:alt: mat_16.png
:width: 400px
Componentes de la imanación en un campo magnético constante según las ecuaciones de Bloch
```
```{figure} ../images/mat_17.png
:alt: mat_17.png
:width: 400px
Evolución de vector imanación en un campo magnético constante según las ecuaciones de Bloch
```


````

**Campo magnético rotatorio $\vec{B} = B_0 \vec{u_z} + B_1 (\cos (\omega t) \vec{u_x} + \sin (\omega t) \vec{u_y})$**

Cuando la excitación es un campo magnético rotatorio, la solución estacionaria de esta ecuación vectorial es la siguiente:

$$
M_x &= M_1 \cos \varphi \cos (\omega t) + M_1 \sin \varphi \sin (\omega t) = M_1 \cos(\omega t - \varphi) \\
M_y &= M_1 \cos \varphi \sin (\omega t) - M_1 \sin \varphi \cos (\omega t) = M_1 \sin(\omega t - \varphi) \\
M_z &= M_0'
$$

Es decir, la imanación sigue a la excitación con un cierto retraso, y mantiene una imanación transversal rotatoria. Obsérvese que la imanación final longitudinal es menor que en ausencia de campo magnético rotatorio, $M_0'<M_0$, y la constante de tiempo de relajación longitudinal $\tau_1'$ es mayor que en ausencia de campo magnético rotatorio.
