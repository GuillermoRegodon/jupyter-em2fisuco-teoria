# Teoremas de conservación

A partir de las ecuaciones de Maxwell podemos estudiar **el efecto que tienen los campos sobre las partículas de un volumen**. Como resultado, nos aparecerán naturalmente *dos términos electromagnéticos* puros: uno *en la superficie* del volumen y otro *en el interior* del volumen. **El efecto se produce a través de la Fuerza de Lorentz**. Dependiendo de si lo que nos interesa es la energía, el momento lineal o el momento angular (las tres magnitudes que se conservan en la mecánica clásica) empezamos el cálculo en un punto diferente:

Para la energía cinética:

* La energía cinética de una partícula:

$$
\frac{d \varepsilon_{c,1}}{dt} = \frac{d}{dt} \left( \frac{1}{2}m v^2 \right) = m \vec{v} \cdot \frac{d \vec{v}}{d t} = m \vec{v} \cdot \vec{a} = \vec{v} \cdot \vec{F} = \vec{v} \cdot \left(q (\vec{E} + \vec{v} \times \vec{B}) \right) = q \vec{v} \cdot \vec{E}
$$

* La energía cinética de todas las partículas de un volumen, sustituyendo $q \rightarrow \rho dV$ e integrando en el volumen $V$.

$$
\frac{d \varepsilon_c}{dt} = \int_V (\rho dV) \vec{v} \cdot \vec{E} = \int_V (\vec{J} \cdot \vec{E}) dV
$$


Para el momento lineal:

* El momento lineal de una partícula:

$$
\frac{d \vec{p_1}}{dt} = \vec{F} = q (\vec{E} + \vec{v} \times \vec{B})
$$

* El momento lineal total de todas las partículas de un volumen, sustituyendo $q \rightarrow \rho dV$ e integrando en el volumen $V$.

$$
\frac{d \vec{p}}{dt} = \int_V (\rho dV) \vec{E} + \vec{v} \times \vec{B} = \int_V (\rho \vec{E} + \vec{J} \times \vec{B}) dV
$$


Respecto al momento angular:

* El momento angular de una partícula, con respecto al origen:

$$
\frac{d \vec{L_1}}{dt} = \vec{r} \times \vec{F} = q \vec{r} \times (\vec{E} + \vec{v} \times \vec{B})
$$

* El momento angular total de todas las partículas de un volumen, sustituyendo $q \rightarrow \rho dV$ e integrando en el volumen $V$.

$$
\frac{d \vec{L}}{dt} = \int_V (\rho dV) \vec{r} \times (\vec{E} + \vec{v} \times \vec{B}) = \int_V \vec{r} \times (\rho \vec{E} + \vec{J} \times \vec{B}) dV
$$


El siguiente paso es sustituir la densidad de carga $\rho$ y la densidad de corriente $\vec{J}$ utilizando las ecuaciones de Maxwell para obtener una expresión en función de los campos. **Este paso es algebraico, se trata tan solo de sustituciones que siempre son válidas, pero en este caso, nos van a permitir identificar términos que tienen sentido físico**. En los tres teoremas de conservación casos vamos a tener tres términos:

* Un término relacionado con **las partículas del volumen**.
* Un término relacionado con **los campos en el interior**.
* Un término que se integra en la superficie y por lo tanto representa **una interacción entre el volumen y el exterior**.

Veamos el caso de la energía, que es menos elaborado algebraicamente:


## Conservación de la energía

Para la energía cinética total de todas las partículas de un volumen obtuvimos:

$$
\frac{d \varepsilon_c}{dt} = \int_V (\vec{J} \cdot \vec{E}) dV
$$

La expresión del interior de la integral se puede desarrollar para obtener una expresión que sólo dependa de los campos, sustituyendo $\vec{J}$ por los campos en la ley de Ampere generalizada.

$$
\vec{J} \cdot \vec{E} = \vec{E} \cdot \left( \nabla \times \vec{H} - \frac{\partial \vec{D}}{\partial t} \right)
$$

Podemos simetrizar la expresión observando que $\vec{E} \cdot \nabla \times \vec{H}$ aparece en el desarrollo de $\nabla \cdot \left(\vec{E} \times \vec{H} \right)$ y usando finalmente la ley de Gauss.

$$
\vec{J} \cdot \vec{E} = - \nabla \cdot \left(\vec{E} \times \vec{H} \right) + \vec{H} \cdot \nabla \times \vec{E} - \vec{E} \cdot \frac{\partial \vec{D}}{\partial t} = - \nabla \cdot \left(\vec{E} \times \vec{H} \right) - \vec{H} \cdot \frac{\partial \vec{B}}{\partial t} - \vec{E} \cdot \frac{\partial \vec{D}}{\partial t}
$$

Esta expresión es la expresión diferencial del teorema de Poynting. El primer término es una divergencia por lo que se puede sustituir la integral de volumen de la divergencia por la integral de superficie (cerrada) del vector. Con esto, obtenemos la expresión integral del teorema de Poynting.

$$
\frac{d \varepsilon_c}{dt} = \int_V (\vec{J} \cdot \vec{E}) dV = - \int_V \left( \vec{H} \cdot \frac{\partial \vec{B}}{\partial t} + \vec{E} \cdot \frac{\partial \vec{D}}{\partial t} \right) - \oint_S \left(\vec{E} \times \vec{H} \right) \cdot \vec{dS}
$$

Con lo que identificamos los términos mencionados anteriormente:

* El miembro de la izquierda, ya sea $\frac{d \varepsilon_c}{dt}$ o su expresión integral $\int_V (\vec{J} \cdot \vec{E}) dV$, **es la variación de energía cinética de todas las partículas del interior** del volumen V.

* El primer término del miembro derecho se integra en el volumen y solo contiene los campos $\vec{E}$ y $\vec{H}$, por lo que podemos identificarlo como **la variación de energía almacenada por los campos $\vec{E}$ y $\vec{H}$ en el interior del volumen**.

* El segundo término del miembro derecho se integra en la superficie y también depende exclusivamente de los campos $\vec{E}$ y $\vec{H}$, por lo que lo identificamos como **el flujo de la energía entre el interior y el exterior del volumen debido a los campos $\vec{E}$ y $\vec{H}$ electromagnéticos**. De hecho, el vector $\vec{G} = \vec{E} \times \vec{H}$ se conoce como **vector de Poynting** y tiene unidades de energía por unidad de tiempo y por unidad de superficie.

```{note}
¡El signo del término de flujo de energía electromagnética es negativo! Esto quiere decir que el flujo se define positivo hacia fuera.
```

## Conservación del momento lineal

El teorema de conservación de la energía tiene la ventaja de ser escalar, por lo que algebraicamente es más sencillo. En particular, el uso del teorema de la divergencia es el uso común que solemos usar en la ley de Gauss. En la conservación del momento lineal, sin embargo, tenemos que hay tres componentes que se van a conservar y que queremos relacionar con los campos $\vec{E}$ y $\vec{H}$, por lo que el uso de la divergencia es menos común, aunque igualmente válido.

````{note}
El teorema de la divergencia escalar dice:

$$
\int_V \left( \nabla \cdot \vec{A} \right) dV = \oint_S \vec{A} \cdot \vec{dS}
$$

Podemos utilizar la siguiente justificación para recordarla utilizando las cordenadas cartesianas.

$$
\int_V \left( \nabla \cdot \vec{A} \right) dV &= \int_V \left( \frac{\partial A_x}{\partial x} + \frac{\partial A_y}{\partial y} + \frac{\partial A_z}{\partial z}\right) dx \space dy \space dz = \\
&= \int_V \left( \frac{\partial A_x}{\partial x} dx \space dy \space dz + \frac{\partial A_y}{\partial y} dx \space dy \space dz + \frac{\partial A_z}{\partial z} dx \space dy \space dz \right) = \\
&= \int_V \left( dA_x \space dy \space dz + dA_y \space dx \space dz + dA_z \space dx \space dy \right)
$$

Cada uno de los tres términos de la integral es un diferencial exacto en $dA_x$, $dA_y$ o $dA_z$. Ésta es la expresión que luego integramos entre un extremo y otro del volumen en cada una de las direcciones para obtener la integral de superficie. No necesitamos terminar la demostración que, de todas formas, es conocida de los cursos de Matemáticas. Por supuesto, el producto $\vec{A} \cdot \vec{dS}$ también es un escalar.

En el teorema de conservación del momento lineal (y en el de conservacion del momento angular) vamos a obtener que los diferenciales exactos son más largos y dependeran de términos de los campos en las tres componentes, y de hecho, utilizando notación tensorial, la expresión queda simetrizada y muy parecida. Sea un tensor $\overset\leftrightarrow{F}$ que podemos usar como una matriz de 9 elementos $F_{ij}$.

$$
\int_V \left( \nabla \cdot \overset\leftrightarrow{F} \right) dV = \oint_S \overset\leftrightarrow{F} \cdot \vec{dS}
$$

Para entender esta expresión, veamos que $\nabla \cdot \vec{A}$ es un escalar, mientras que $\nabla \cdot \overset\leftrightarrow{F}$ es un vector, o más bien la suma de tres vectores:

$$
\nabla \cdot \overset\leftrightarrow{F} = \frac{\partial (F_{xx}, F_{xy}, F_{xz})}{\partial x} + \frac{\partial (F_{yx}, F_{yy}, F_{yz})}{\partial y} + \frac{\partial (F_{zx}, F_{zy}, F_{zz})}{\partial z}
$$

Por lo demás, el teorema de la divergencia funciona igual: al final obtenemos diferenciales exactos (de vectores) que podremos integrar entre los extremos para convertir en una integral de superficie cerrada. Veamos por último que $\overset\leftrightarrow{F} \cdot \vec{dS}$ también es un vector. Usando el producto matricial:

$$
\overset\leftrightarrow{F} \cdot \vec{dS} = \left( \begin{array}{ccc} F_{xx} & F_{xy} & F_{xz} \\ F_{yx} & F_{yy} & F_{yz} \\ F_{zx} & F_{zy} & F_{zz} \end{array} \right) \left( \begin{array}{c} dS_x \\ dS_y \\ dS_z \end{array} \right) = \left( \begin{array}{c} F_{xx}dS_x + F_{xy}dS_y + F_{xz}dS_z \\ F_{yx}dS_x + F_{yy}dS_y + F_{yz}dS_z \\ F_{zx}dS_x + F_{zy}dS_y + F_{zz}dS_z \end{array} \right) 
$$

Con lo que la integral de superficie $\oint_S \overset\leftrightarrow{F} \cdot \vec{dS}$ es un vector con tres integrales de superficie.

```{note}
Estas expresiones quedarán más sencillas usando la notación tensorial pura, de forma que aplicar el teorema de la divergencia a un tensor de orden 2 no genera ninguna dificultad.
```
````

Para el momento lineal total de todas las partículas de un volumen obtuvimos:

$$
\frac{d \vec{p}}{dt} = \int_V (\rho \vec{E} + \vec{J} \times \vec{B}) dV
$$

Un desarrollo algo elaborado nos permite obtener finalmente la expresión integral del teorema de conservación del momento lineal.

$$
\frac{d \vec{p}}{dt} &= - \frac{d}{dt} \int_V \varepsilon \mu \left( \vec{E} \times \vec{H} \right) dV + \int_V \left( \left( \vec{E} (\nabla \cdot \vec{D}) - \vec{D} \times (\nabla \times \vec{E}) \right) + \left( \vec{H} (\nabla \cdot \vec{B}) - \vec{B} \times (\nabla \times \vec{H}) \right) \right) dV \\
&= - \frac{d}{dt} \int_V \varepsilon \mu \left( \vec{E} \times \vec{H} \right) dV + \int_V \nabla \cdot \overset\leftrightarrow{T} dV \\
&= - \frac{d}{dt} \int_V \varepsilon \mu \left( \vec{E} \times \vec{H} \right) dV + \oint_S \overset\leftrightarrow{T} \cdot \vec{dS}
$$

Identificamos los términos mencionados anteriormente:

* El miembro de la izquierda, ya sea $\frac{d \vec{p}}{dt}$ o su expresión integral $\int_V (\rho \vec{E} + \vec{J} \times \vec{B}) dV$, **es la variación de momento lineal de todas las partículas del interior** del volumen.

* El primer término del miembro derecho se integra en el volumen y solo contiene los campos $\vec{E}$ y $\vec{H}$, por lo que podemos identificarlo como **la variación de momento lineal almacenado por los campos $\vec{E}$ y $\vec{H}$ en el interior del volumen**. De hecho, el vector $\vec{g} = \varepsilon \mu \vec{E} \times \vec{H}$ es la densidad de momento lineal electromagnético y tiene unidades de momento lineal por unidad de volumen.

* El segundo término del miembro derecho se integra en la superficie y también depende exclusivamente de los campos $\vec{E}$ y $\vec{H}$, por lo que lo identificamos como **el flujo de momento lineal entre el interior y el exterior del volumen debido a los campos $\vec{E}$ y $\vec{H}$ electromagnéticos**. Tiene el signo positivo, a diferencia del término equivalente en el teorema de Poynting. El integrando es el conocido como tensor de esfuerzos de Maxwell y se puede escribir de las siguientes maneras equivalentes:

$$
\nabla \cdot \overset\leftrightarrow{T} &= \left( \vec{E} (\nabla \cdot \vec{D}) - \vec{D} \times (\nabla \times \vec{E} ) \right) + \left( \vec{H} (\nabla \cdot \vec{B}) - \vec{B} \times (\nabla \times \vec{H}) \right) \\
T_{ij} &= E_iD_j + H_iB_j - \frac{1}{2}\left( E_kD_k + H_kB_k \right) \delta_{ij} \quad (\sum_k \rm {implicita}) \\
\overset\leftrightarrow{T} &= \vec{E}\vec{D} + \vec{H}\vec{B} - \frac{1}{2} \left( \vec{E} \cdot \vec{D} + \vec{H} \cdot \vec{B} \right) \overset\leftrightarrow{I} \quad (\overset\leftrightarrow{I} \rm{\space tensor \space diagonal})
$$

```{note}
¡El signo del término de flujo de momento lineal es positivo! Esto quiere decir que el flujo se define positivo hacia dentro.
```


## Conservación del momento angular

Se puede llegar a una expresión similar, empezando desde la expresión que obtuvimos para el momento angular de todas las partículas de un volumen:

$$
\frac{d \vec{L}}{dt} = \int_V \vec{r} \times (\rho \vec{E} + \vec{J} \times \vec{B}) dV
$$

Similarmente, llegaríamos a la expresión integral de la conservación del momento angular.

$$
\frac{d \vec{L}}{dt} = - \frac{d}{dt} \int_V \left( \vec{r} \times \left( \varepsilon \mu ( \vec{E} \times \vec{H} ) \right) \right) dV + \oint_S \left( \vec{r} \times \overset\leftrightarrow{T} \right) \cdot \vec{dS}
$$

Igualmente, identificamos los términos mencionados anteriormente:

* El miembro de la izquierda, ya sea $\frac{d \vec{L}}{dt}$ o su expresión integral $\int_V \vec{r} \times (\rho \vec{E} + \vec{J} \times \vec{B}) dV$, **es la variación de momento angular de todas las partículas del interior** del volumen.

* El primer término del miembro derecho se integra en el volumen y solo contiene los campos $\vec{E}$ y $\vec{H}$, por lo que podemos identificarlo como **la variación de momento angular almacenado por los campos $\vec{E}$ y $\vec{H}$ en el interior del volumen**. De hecho, el vector $\vec{r} \times \vec{g} = \vec{r} \times \left( \varepsilon \mu \vec{E} \times \vec{H} \right)$ es la densidad de momento angular electromagnético y tiene unidades de momento angular por unidad de volumen.

* El segundo término del miembro derecho se integra en la superficie y también depende exclusivamente de los campos $\vec{E}$ y $\vec{H}$, por lo que lo identificamos como **el flujo de momento angular entre el interior y el exterior del volumen debido a los campos $\vec{E}$ y $\vec{H}$ electromagnéticos**. Tiene el signo positivo.

```{note}
¡El signo del término de flujo de momento angular es positivo! Esto quiere decir que el flujo se define positivo hacia dentro.
```
