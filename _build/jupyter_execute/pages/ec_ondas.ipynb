# Ondas electromagnéticas

Vamos a repasar los conceptos matemáticos que utilizamos para estudiar las ondas electromagnéticas de forma más ventajosa.

## Ecuación de ondas 1D y 3D con dirección arbitraria

**La ecuación de ondas 1D** es la siguiente:

$$
\frac{∂^2 f}{∂x^2} - \frac{1}{c^2} \frac{∂^2 f}{∂t^2}=0
$$

Y **tiene por solución** la siguiente función genérica.

$$
f(x,t) = f_1(x-ct) + f_2(x+ct)
$$

Es decir, la solución es:
* cualquier **función $f_1$** (*suficientemente derivable*) **que se desplaza a la derecha** ($x$ creciente) a medida que el tiempo avanza (cuando $t$ aumenta), más
* otra **función $f_2$** (*suficientemente derivable*) **que se desplaza a la izquierda** ($x$ decreciente) a medida que el tiempo avanza (cuando $t$ aumenta).

Estas funciones son las que llamamos **ondas progresiva y regresiva**.

```{note}
Podemos comprobar que es solución por simple sustitución utilizando la regla de la cadena:

$$
\frac{∂f_i}{∂x} &= f_i' &\quad &\frac{∂^2f_i}{∂x^2} = f_i'' \\
\frac{∂f_i}{∂t} &= \pm c f_i' &\quad &\frac{∂^2f_i}{∂t^2} = c^2 f_i''
$$

(*$f_i'$ y $f_i'$ son las derivadas con respecto a su argumento*) Podemos demostrar además que es la única solución posible. Para ello, definimos las variables $\chi = x- ct$ y $\eta = x + ct$. Sabemos que la única solución posible para $\frac{∂f}{∂\chi} = 0$ es una función genérica que no depende de $\chi$, o lo que es lo mismo, una función $f_2(\eta)$ que solo depende de $\eta$. Igualmente, $\frac{∂f}{∂\eta} = 0$ tiene por solución $f_1(\chi)$. La ecuación segunda derivada siguiente tiene por única solución la suma de las dos:

$$
\frac{∂^2 f}{∂\chi ∂\eta} = 0 \quad \rightarrow \quad f = f_1(\chi) + f_2(\eta)
$$

Solo nos queda verificar por la regla de la cadena que $\frac{∂^2 f}{∂\chi ∂\eta} = 0$ es la ecuación de onda:

$$
\frac{∂f}{∂\chi} = \frac{∂f}{∂x}\frac{∂x}{∂\chi} + \frac{∂f}{∂t}\frac{∂t}{∂\chi} = \frac{∂f}{∂x} -\frac{1}{c} \frac{∂f}{∂t} \\
$$

$$
\frac{∂g}{∂\eta} = \frac{∂g}{∂x}\frac{∂x}{∂\eta} + \frac{∂g}{∂t}\frac{∂t}{∂\eta} = \frac{∂g}{∂x} +\frac{1}{c} \frac{∂g}{∂t}
$$

Ponemos $g$ en la segunda ecuación eligiendo $g = \frac{∂f}{∂\chi}$ y hemos demostrado que son la misma ecuación. Por lo tanto, si $f = f_1(\chi) + f_2(\eta)$ es la única solución de la ecuación $\frac{∂^2 f}{∂\chi ∂\eta} = 0$, entonces $f = f_1(x-ct) + f_2(x+ct)$ es la única solución de la ecuación de onda 1D.

```

**La ecuación de ondas 3D** es la siguiente:

$$
\nabla^2 f - \frac{1}{c^2} \frac{∂^2 f}{∂t^2}=0
$$

La solución 3D a priori no está relacionada con la 1D, pero **vamos a ver que la relación entre ambas ecuaciones es directa si elegimos una dirección preferente**, indicada por un vector $\vec{n}$ cualquiera (*al que imponemos $|\vec{n}| = 1$ por simplicidad en las expresiones posteriores*), y elegimos que la función $f$ dependa de las variables espaciales relacionadas según:

$$
\tau = \vec{n} \cdot \vec{r} = n_xx + n_yy + n_zz
$$

Es decir, $f(x,y,z,t) = f(\tau,t)$. De esta manera, el Laplaciano pasa a ser $\nabla^2 f = \frac{∂^2f}{∂\tau^2}$ y **la ecuación de ondas 3D se convierte en una ecuación de onda 1D en cualquier dirección** dada por $\vec{n}$:

$$
\frac{∂^2f}{∂\tau^2} - \frac{1}{c^2} \frac{∂^2 f}{∂t^2}=0
$$

Como la dirección es arbitraria, y la ecuación de ondas es lineal, la suma de soluciones en distintas direcciones también es solución. Es decir, **la solución de la ecuación de ondas 3D es la suma de cualquier número de ondas 1D en cualquier dirección del espacio**. Dado que la onda regresiva en la dirección $\vec{n}$ es igual a sustituir $\vec{n} \rightarrow -\vec{n}$, podemos olvidarnos de las ondas regresivas hasta que nos haga falta.

```{note}
La demostración para pasar de la ecuación de ondas 3D a la ecuación de ondas 1D es inmediata usando la regla de la cadena. Usando $\tau = n_xx + n_yy + n_zz$:

$$
\frac{∂f}{∂x} &= \frac{∂f}{∂\tau}\frac{∂\tau}{∂x} = n_x\frac{∂f}{∂\tau} \\
\frac{∂^2f}{∂x^2} &= n_x^2\frac{∂^2f}{∂\tau^2} \\
\nabla^2 f &= \frac{∂^2f}{∂x^2} + \frac{∂^2f}{∂y^2} + \frac{∂^2f}{∂z^2} = (n_x^2 + n_y^2 + n_z^2) \frac{∂^2f}{∂\tau^2} = \frac{∂^2f}{∂\tau^2}
$$

(*Por este motivo elegimos $\vec{n}$ normalizado, $|\vec{n}|=1$*)
```

````{note}
Veamos la interpretación geométrica de $\tau$ y de $\vec{n}$. El vector $\vec{n}$ es un vector unitario arbitrario. Es más fácil probar los vectores unitarios en la dirección de los ejes. Si $\vec{n} = \vec{u_x}$:

$$
\tau = \vec{n} \cdot \vec{r} = \vec{u_x} \cdot \vec{r} = x
$$

Es decir, estamos hablando de los puntos que comparten la misma componente $x = \tau$, o lo que es lo mismo, planos perpendiculares al vector $\vec{u_x}$ que se desplazan hacia $+x$ cuando $\tau$ aumenta. De la misma manera, los puntos $\vec{r}$ que cumplen la expresión $\tau = \vec{n} \cdot \vec{r}$ son los que comparten la misma componente en la dirección de $\vec{n}$, y además dicha componente tiene una longitud $\tau$ o, dicho de otro modo, $\tau = \vec{n} \cdot \vec{r}$ son planos que se desplazan en la dirección $\vec{n}$ a medida que $\tau$ aumenta.

````



## Fasores

La solución de la ecuación de onda 3D es una función cualquiera que se desplaza en cualquier dirección (*nos podemos olvidar de la onda regresiva, hasta que nos haga falta*). Del análisis de Fourier, sabemos que cualquier función se puede representar en función de sus componentes armónicas a cada frecuencia $\omega$, y como la ecuación de ondas es lineal, **si cada una de las componentes armónicas cumple la ecuación de ondas 3D, la composición también la cumple y es solución**.

```{note}
Por este motivo, **Podemos estudiar tan solo una componente armónica de cualquier frecuencia $\omega$ de forma individual**.
```

La expresión de la componente armónica es:

$$
f = A \cos(\frac{\omega}{c}(\tau-ct)-\varphi)
$$

Donde $A$ y $\varphi$ son constantes. Elegimos que el argumento esté multiplicado por $\frac{\omega}{c}$ para poder poner el argumento del coseno de la forma más común, definiendo además la constante $k = \frac{\omega}{c}$.

$$
\frac{\omega}{c}(\tau-ct) = k\tau - \omega t
$$

De hecho, por costumbre, como el coseno es una función par, se suele cambiar el signo también:

$$
f = A \cos(\omega t - k\tau + \varphi)
$$

```{note}
Observemos que la velocidad de la onda $c$ es cualquier número pero tiene que ser el que aparece en la ecuación de ondas, es decir que es único para la ecuación.

$$
\nabla^2 f - \frac{1}{c^2} \frac{∂^2 f}{∂t^2}=0
$$

La frecuencia $\omega$ es arbitraria, así que puede haber ondas con cualquier frecuencia $\omega$, pero todas a la misma velocidad, $c$. El número de onda $k$ se define como una magnitud derivada de la frecuencia y la velocidad.

$$
k =\frac{\omega}{c}
$$
```

Además, **la solución es una función real pero la ecuación de ondas es perfectamente válida para números complejos**, así que podemos trabajar con una exponencial compleja y tomar la parte real. Y de hecho, veremos que típicamente hablaremos simplemente del módulo de la exponencial, que será el número complejo que multiplica a la exponencial, $A$.

$$
f(\tau,t) = A_0 \cos(\omega t - k \tau + \varphi) = \Re(A_0 e^{j \varphi} e^{j(\omega t - k \tau)}) = \Re(A e^{j(\omega t - k \tau)})
$$

La ventaja de usar exponenciales es que las derivadas se pueden sustituir por productos complejos.

$$
\frac{∂}{∂\tau} \left( A e^{j(\omega t - k \tau)} \right) &= - j k A e^{j(\omega t - k \tau)} \\ 
\frac{∂}{∂t} \left( A e^{j(\omega t - k \tau)} \right) &= j \omega A e^{j(\omega t - k \tau)}
$$

````{note}
Se puede demostrar la validez de usar tan solo el fasor en ecuaciones que tengan las siguientes propiedades:

* La solución debe ser real.
* Los coeficientes de la función y sus derivadas son reales.
* La ecuación es lineal.

La ecuación de ondas cumple estas propiedades:

$$
\nabla^2 f - \frac{1}{c^2} \frac{∂^2 f}{∂t^2}=0
$$

* La solución es una magnitud física, y por lo tanto debe ser real.
* Los coeficientes son $1$ o $\frac{1}{c^2}$.
* Si $f_1$ y $f_2$ son soluciones, entonces $f_1 + f_2$ también es solución.

Veamos por qué se puede hacer, de una forma sencilla. El coseno se puede poner en función de dos exponenciales complejas:

$$
A_0 \cos(\omega t - k \tau + \varphi) = \frac{A_0}{2} e^{j \varphi} e^{j(\omega t - k \tau)} + \frac{A_0}{2} e^{-j \varphi} e^{-j(\omega t - k \tau)} = \frac{A}{2} e^{j(\omega t - k \tau)} + \frac{\bar{A}}{2}  e^{-j(\omega t - k \tau)}
$$

Por lo tanto: 

1. Resolvemos la ecuación para la exponencial compleja positiva: esto es más facil si podemos sustituir las derivadas por productos complejos, lo que nos dará facilmente una solución $A = A_0 e^{j\varphi}$.
2. La exponencial compleja negativa es automáticamente solución: solo tenemos que tomar complejo conjugado en la ecuación a resolver para comprobar que $\bar{A} = A_0 e^{-j\varphi}$ también es solución.
3. La suma de ambas soluciones también es solución, así que el coseno es solución.

```{note}
**Si la ecuación en fasores tiene solución, entonces la onda armónica que construimos a partir de dicho fasor, también es solución**.
```
````

## Ecuaciones de Maxwell en fasores

Elegimos una dirección arbitraria $\vec{n}$ para la onda, de tal modo que $|\vec{n}| = 1$ y $\tau = \vec{n} \cdot \vec{r}$, y buscamos una solución multiplicada por una función exponencial $e^{j(\omega t - k\tau)}$. **Las ecuaciones de Maxwell son ecuaciones vectoriales: La solución es un par de vectores $\vec{E}$ y $\vec{B}$**. Entonces, la solución que buscamos es:

$$
\vec{E}(\tau,t) = \vec{E} e^{j(\omega t - k\tau)} \\
\vec{B}(\tau,t) = \vec{B} e^{j(\omega t - k\tau)}
$$

El operador $\nabla$ que aparece en las ecuaciones de Maxwell se puede poner de la siguiente manera:

$$
\nabla = \left( \frac{∂}{∂x}, \frac{∂}{∂y}, \frac{∂}{∂z} \right) = \left( \frac{∂\tau}{∂x}\frac{∂}{∂\tau}, \frac{∂\tau}{∂y}\frac{∂}{∂\tau}, \frac{∂\tau}{∂z}\frac{∂}{∂\tau} \right) = \left( n_x, n_y, n_z \right) \frac{∂}{∂\tau} = \vec{n}\frac{∂}{∂\tau}
$$

Para campos eléctricos y magnéticos con esta forma, la divergencia y el rotacional que pueden poner de la siguiente manera (*$\vec{F}$ puede ser $\vec{E}$ o $\vec{B}$*):

$$
\nabla \cdot \vec{F}(\tau,t) &= (\vec{n} \frac{∂}{∂\tau}) \cdot \vec{F} = \vec{n} \cdot \frac{∂\vec{F}}{∂\tau} = -j k \vec{n} \cdot \vec{F}(\tau,t) \\
\nabla \times \vec{F}(\tau,t) &= (\vec{n} \frac{∂}{∂\tau}) \times \vec{F} =  \vec{n} \times \frac{∂ \vec{F}}{∂\tau} = -j k  \vec{n} \times \vec{F}(\tau,t)
$$

Por lo que **las ecuaciones de Maxwell en un medio isótropo, sin cargas libres $\rho_l$ ni corrientes de conducción $\vec{J_c}$, para ondas armónicas en la dirección arbitraria $\vec{n}$**, son, antes de sustituir las derivadas por productos complejos:

$$
\vec{n} \cdot \frac{∂\vec{E}(\tau,t)}{∂\tau} &= 0 \\
\vec{n} \cdot \frac{∂\vec{B}(\tau,t)}{∂\tau} &= 0 \\
\vec{n} \times \frac{∂\vec{E}(\tau,t)}{∂\tau} &= - \frac{∂\vec{B}(\tau,t)}{∂t} \\
\vec{n} \times \frac{∂\vec{B}(\tau,t)}{∂\tau} &= \epsilon \mu \frac{∂\vec{E}(\tau,t)}{∂t} \\
$$

Y, después de sustituir las derivadas por productos complejos, tenemos **la ecuación en fasores**:

$$
-j k \vec{n} \cdot \vec{E} &= 0 \\
-j k \vec{n} \cdot \vec{B} &= 0 \\
-j k \vec{n} \times \vec{E} &= - j \omega \vec{B} \\
-j k \vec{n} \times \vec{B} &= \epsilon \mu (j \omega \vec{E})
$$

De donde obtenemos finalmente **todas las propiedades que cumplen los fasores del campo eléctrico y magnético en una onda electromagnética plana**.

$$
\vec{n} \cdot \vec{E} &= \vec{n} \cdot \vec{B} = 0 \\
\vec{n} \times \vec{E} &= c \vec{B} \\
\vec{n} \times \vec{B} &= \frac{1}{c} \vec{E}
$$

```{note}
**Los fasores del campo eléctrico $\vec{E}$ y magnético $\vec{B}$ en una onda electromagnética plana son perpendiculares a la dirección de propagación $\vec{n}$ y perpendiculares entre sí**.
```
