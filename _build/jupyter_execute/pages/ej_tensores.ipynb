# Ejemplos de aplicación

Vamos a repasar algunos ejemplos para verificar las relaciones tensoriales obtenidas.

## Diferencial de área en polares

El diferencial de área es el trozo de área que obtenemos cuando variamos diferencialmente las coordenadas. Para ello, variamos el vector de posición con respecto a una de las coordenadas y lo multiplicamos por su diferencial $dx^i\frac{\partial \vec{r}}{\partial x^i} = dx^i \vec{e_i}$, luego hacemos lo propio con otra coordenada $x^j \vec{e_j}$, y los multiplicamos vectorialmente para obtener el diferencial de área en el plano caracterizado por las coordenadas $x^i$ y $x^j$. Si no queremos conservar la dirección, tomamos el módulo del vector $\vec{dA}$ obtenido:

$$
dA = |dx^i \vec{e_i} \times dx^j \vec{e_j}| \quad \text{(sin suma en $i$ o $j$)}
$$

```{figure} ../images/tensores_7.png
:alt: tensores_7.png
:width: 325px
Diferencial de área en cualquier sistema de coordenadas
```

En cartesianas, nos queda:

$$
dA = |dx \space \vec{u_x} \times dy \space \vec{u_y}| = dx \space dy
$$

En polares, hemos visto que $\vec{e_r} = \vec{u_r}$ y $\vec{e_\theta} = r\vec{u_\theta}$, por lo que nos queda:

$$
dA = |dr \space \vec{e_r} \times d\theta \space \vec{e_\theta}| = |dr \space \vec{u_r} \times d\theta \space r\vec{u_\theta}| = r\space dr \space d\theta
$$


## Gradiente en polares

El gradiente en cartesianas sigue la expresión:

$$
\nabla = (\frac{\partial}{\partial x}, \frac{\partial}{\partial y}) = \vec{u_x} \frac{\partial}{\partial x} + \vec{u_y} \frac{\partial}{\partial y} = \frac{\partial \vec{r}}{\partial x} \frac{\partial}{\partial x} + \frac{\partial \vec{r}}{\partial y} \frac{\partial}{\partial y}
$$

Observamos que cada uno de los elementos se transforma de forma covariante, tanto los vectores de la base natural como las derivadas parciales. Si transformamos cada uno de los elementos a polares, obtenemos:

$$
\nabla = \left( \frac{\partial \vec{r}}{\partial r} \frac{\partial r}{\partial x} + \frac{\partial \vec{r}}{\partial \theta} \frac{\partial \theta}{\partial x} \right) \left( \frac{\partial}{\partial r} \frac{\partial r}{\partial x} + \frac{\partial}{\partial \theta} \frac{\partial \theta}{\partial x} \right) + \left( \frac{\partial \vec{r}}{\partial r} \frac{\partial r}{\partial y} + \frac{\partial \vec{r}}{\partial \theta} \frac{\partial \theta}{\partial y} \right) \left( \frac{\partial}{\partial r} \frac{\partial r}{\partial y} + \frac{\partial}{\partial \theta} \frac{\partial \theta}{\partial y} \right)
$$

Debemos usar $\vec{r} = (r\cos\theta , r\sin\theta)$ y las parciales de cambio de coordenadas. Realizando el cálculo de forma detallada, nos queda:

$$
\nabla = \vec{u_r} \frac{\partial}{\partial r} + \frac{1}{r} \vec{u_\theta} \frac{\partial}{\partial \theta}
$$


## Rozamiento en notación tensorial

Vamos a ver un ejemplo que se suele resolver por equilibrio de fuerzas en un sistema de referencia elegido convenientemente.

```{figure} ../images/tensores_8.png
:alt: tensores_8.png
:width: 325px
Problema de rozamiento.
```

Se elije el eje $x$ de forma que la fuerza de rozamiento tenga solo componente $\vec{F_r} = F_{rx}\vec{u_x}$ (*ver que $F_{rx}$ es negativo*). La fuerza normal $\vec{N}$ tiene solo componente $y$ y debe compensar la componente $y$ de la resultante de la fuerza de arrastre $\vec{F_a}$ y el peso $\vec{P}$ La ley de la fuerza de rozamiento es la siguiente:

$$
F_r = F_{rx} = -\mu N
$$

De forma que las ecuaciones del problema son:

$$
F_{rx} &= -\mu (P - F_{ay}) \\
F_{ry} &= 0
$$

Podemos poner estas ecuaciones en forma tensorial independiente del sistema de referencia utilizando las definiciones de $F^i = \vec{F_r}$, $R^i = \vec{F_a} - \vec{P}$ y definiendo el tensor $\mu_i^j$ que los relaciona:

$$
F^j = \mu_i^j R^i
$$

O, en notación matricial:

$$
\left( \begin{array}{c} F_{rx} \\ F_{ry} \end{array} \right) = \left( \begin{array}{cc} 0 & \mu \\ 0 & 0 \end{array} \right) \left( \begin{array}{c} F_{ax} \\ F_{ay} - P \end{array} \right)
$$

La expresión tensorial $F^j = \mu_i^j R^i$ es independiente del sistema de coordenadas, por lo que podemos elegir cualquier sistema de coordenadas y deberíamos obtener las misma expresiones.

````{note}
Para comprobarlo, vamos a probar primero un cambio de coordenadas casi trivial:

$$
x' &= x \\
y' &= -y
$$

```{figure} ../images/tensores_9.png
:alt: tensores_9.png
:width: 325px
Problema de rozamiento.
```

La transformación de los vectores es trivial, pero aún así, vamos a hacer definiendo las parciales de la transformación:

$$
\left( \begin{array}{cc} \frac{\partial x}{\partial x'} & \frac{\partial x}{\partial y'} \\ \frac{\partial y}{\partial x'} & \frac{\partial y}{\partial y'} \end{array} \right) &= \left( \begin{array}{cc} 1 & 0 \\ 0 & -1 \end{array} \right)\\
\left( \begin{array}{cc} \frac{\partial x'}{\partial x} & \frac{\partial x'}{\partial y} \\ \frac{\partial y'}{\partial x} & \frac{\partial y'}{\partial y} \end{array} \right) &= \left( \begin{array}{cc} 1 & 0 \\ 0 & -1 \end{array} \right)
$$

Las componentes de los vectores se transforman de forma contravariante, es decir usando $\frac{\partial x'^i}{\partial x^j}$.

$$
F'^i &= \frac{\partial x'^i}{\partial x^j} F^j = \left( \begin{array}{c} F_{rx} \\ -F_{ry} \end{array} \right) \\
R'^i &= \frac{\partial x'^i}{\partial x^j} R^j = \left( \begin{array}{c} F_{ax} \\ P-F_{ay} \end{array} \right) \\
$$

Mientras que el tensor $\mu_i^j$ se transforma de la siguiente manera:

$$
\mu_i'^j = \frac{\partial x'^j}{\partial x^k}\frac{\partial x^l}{\partial x'^i} \mu_l^k = \left( \begin{array}{cc} 0 & -\mu \\ 0 & 0 \end{array} \right)
$$

De forma que la expresión tensorial $F'^j = \mu_i'^j R'^i$ sigue siendo válida en el nuevo sistema de coordenadas:

$$
\left( \begin{array}{c} F_{rx} \\ -F_{ry} \end{array} \right) = \left( \begin{array}{cc} 0 & -\mu \\ 0 & 0 \end{array} \right) \left( \begin{array}{c} F_{ax} \\ P - F_{ay} \end{array} \right)
$$

````

````{note}
Vamos a hacer lo mismo, pero en un sistema de coordenadas no tan trivial, una rotación:

$$
x' &= \frac{x+y}{\sqrt{2}} \\
y' &= \frac{-x+y}{\sqrt{2}}
$$

```{figure} ../images/tensores_10.png
:alt: tensores_10.png
:width: 325px
Problema de rozamiento.
```

La transformación inversa, despejando $x$ e $y$:

$$
x &= \frac{x'-y'}{\sqrt{2}} \\
y &= \frac{x'+y'}{\sqrt{2}}
$$

La transformación de los vectores es:

$$
\left( \begin{array}{cc} \frac{\partial x}{\partial x'} & \frac{\partial x}{\partial y'} \\ \frac{\partial y}{\partial x'} & \frac{\partial y}{\partial y'} \end{array} \right) &= \left( \begin{array}{cc} \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{array} \right)\\
\left( \begin{array}{cc} \frac{\partial x'}{\partial x} & \frac{\partial x'}{\partial y} \\ \frac{\partial y'}{\partial x} & \frac{\partial y'}{\partial y} \end{array} \right) &= \left( \begin{array}{cc} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{array} \right)
$$

Las componentes de los vectores se transforman de forma contravariante, es decir usando $\frac{\partial x'^i}{\partial x^j}$.

$$
F'^i &= \frac{\partial x'^i}{\partial x^j} F^j = \left( \begin{array}{c} \frac{F_{rx}+F_{ry}}{\sqrt{2}} \\ \frac{-F_{rx}+F_{ry}}{\sqrt{2}} \end{array} \right) \\
R'^i &= \frac{\partial x'^i}{\partial x^j} R^j = \left( \begin{array}{c} \frac{F_{ax}+F_{ay}-P}{\sqrt{2}} \\ \frac{-F_{ax}-P+F_{ay}}{\sqrt{2}} \end{array} \right) \\
$$

Mientras que el tensor $\mu_i^j$ se transforma de la siguiente manera:

$$
\mu_i'^j = \frac{\partial x'^j}{\partial x^k}\frac{\partial x^l}{\partial x'^i} \mu_l^k = \left( \begin{array}{cc} \frac{\mu}{2} & \frac{\mu}{2} \\ -\frac{\mu}{2} & -\frac{\mu}{2} \end{array} \right)
$$

De forma que la expresión tensorial $F'^j = \mu_i'^j R'^i$ sigue siendo válida en el nuevo sistema de coordenadas:

$$
\left( \begin{array}{c} \frac{F_{rx}+F_{ry}}{\sqrt{2}} \\ \frac{-F_{rx}+F_{ry}}{\sqrt{2}} \end{array} \right) = \left( \begin{array}{cc} \frac{\mu}{2} & \frac{\mu}{2} \\ -\frac{\mu}{2} & -\frac{\mu}{2} \end{array} \right) \left( \begin{array}{c} \frac{F_{ax}+F_{ay}-P}{\sqrt{2}} \\ \frac{-F_{ax}-P+F_{ay}}{\sqrt{2}} \end{array} \right)
$$

````

Ambos ejemplos son sistemas de coordenadas "planos", es decir, el cambio es igual en todo el espacio. Por eso, en los ejemplos la transformación de las coordenadas y de los vectores es la misma. Esto ocurre también el las transformaciones de la relatividad especial.

Sin embargo, si hubiéramos usado un sistema de coordenadas curvilíneo como las polares, los vectores de la base natural, y las parciales de la transformación dependerían de las coordenadas del punto en el que se están aplicando los vectores. Esto es lo que ocurre en la relatividad general.
