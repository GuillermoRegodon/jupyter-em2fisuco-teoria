# Repaso del uso de las simetrías para resolver problemas

Las ecuaciones generales del electromagnetismo en situación estática o cuasiestática tienen una dependencia con el espacio tridimensional: La dependencia temporal se ignora. Sin embargo, si la simetría es apropiada, se puede reducir el orden de las ecuaciones, para que dependan de una sola variable espacial, por lo que la solución resulta ser inmediata. La mayor dificultad está en entender bien los límites de validez y los límites de integración.


## Electrostática: Ley de Gauss

En electrostática, la ecuación que nos permite aprovechar la simetría es la de la ley de Gauss en forma integral:

$$
\int_S \vec{D} \cdot \vec{dS} = Q_{enc}
$$

O en su versión para permitividad constante:

$$
\int_S \vec{E} \cdot \vec{dS} = \frac{Q_{enc}}{\varepsilon}
$$

En el caso del vacio, simplemente usamos la permitividad eléctrica del vacio:

$$
\int_S \vec{E} \cdot \vec{dS} = \frac{Q_{enc}}{\varepsilon_0}
$$

Recordemos que:

$$
Q_{enc} = \int_V\rho dV
$$

**¡Esta ecuación es válida siempre!** Es decir, cualquier superficie cerrada $S$ se puede usar para realizar la integral y sumar la carga total en el volumen interior a dicha superficie cerrada. *Es fundamental saber elegir la superficie que facilita y permite resolver el problema*:

* El objetivo es que la integral $\int_S \vec{D} \cdot \vec{dS}$ se convierta en un producto, por lo que **es fundamental que el vector desplazamiento $\vec{D}$ sea constante**. O, en el caso de permitividad constante, **el campo eléctrico $\vec{E}$ debe ser constante**.

* Además, para que el producto escalar de la integral $\int_S \vec{D} \cdot \vec{dS}$ sea sencillo, **debemos elegir una superficie perpendicular al vector desplazamiento $\vec{D}$ o al campo eléctrico $\vec{E}$**.

* Puede haber otras superficies, *pero deben ser perpendiculares al vector que se integra, para que su contribución sea nula*.

* **La carga en el interior debe ser conocida**.

En problemas de electrostática, solo podemos tener 3 simetrías que cumplen estas condiciones:

### Simetría esférica

Calculamos, por ejemplo, el campo eléctrico generado por una esfera hueca de carga con densidad superficial de carga $\sigma$ en el vacio. La simetría esférica implica que las superficies que debemos elegir son esferas.

* El campo eléctrico es radial $\vec{E} = E_r\vec{u_r}$.
* El diferencial de superficie tambien es radial $\vec{dS} = dS \vec{u_r}$.
* La superficie total de la superficie de Gauss es $4 \pi r^2$.

La esfera $S$ puede tener un radio menor o mayor que el radio $a$ de la esfera de carga. Para radio $r<a$:

```{figure} ../images/gauss_esf_menor.png
:alt: gauss_esf_menor.png
:width: 250px
r < a.
```

La carga encerrada es 0, ya que la esfera es hueca, y la integral se convierte en el producto del campo eléctrico radial por la superficie:

$$
E_r 4 \pi r^2 = 0
$$

Para radio $r>a$:

```{figure} ../images/gauss_esf_mayor.png
:alt: gauss_esf_mayor.png
:width: 250px
r > a.
```

La carga encerrada es la carga total $Q_{enc} = \sigma S = \sigma 4 \pi a^2$:

$$
E_r 4 \pi r^2 = \frac{\sigma 4 \pi a^2}{\varepsilon_0}
$$

### Simetría cilíndrica

Calculamos, por ejemplo, el campo eléctrico generado por un cilindro de densidad de carga $\rho$ y permitividad eléctrica $\varepsilon$ en el interior (el cilindro es no hueco) y vacio en el exterior. Los superficies que cumplen la simetría son cilindros de altura $H$:

* El vector desplazamiento debe ser radial $\vec{D} = D_s\vec{u_s}$.
* El diferencial de superficie lateral es $dS_l = s \space ds \space dz \vec{u_s}$.
* Tenemos que cerrar el cilindro, lo que se puede hacer con tapas perpendiculares al eje. De esta forma, el diferencial de superficie es perpendicular al campo eléctrico y su contribución es nula (nos nos importa su área):
* La cara lateral tiene un área total $S_l = H 2 \pi s$

$$
\int_S \vec{D} \cdot \vec{dS} = \int_{S_l} \vec{D} \cdot \vec{dS_l} + \int_{S_{t1}} \vec{D} \cdot \vec{dS_{t1}} + \int_{S_{t2}} \vec{D} \cdot \vec{dS_{t2}} = \int_{S_l} \vec{D} \cdot \vec{dS_l}
$$

* La carga es la integral de volumen de la densidad de carga, y se puede calcular eligiendo los límites de integración correctamente.

Para radio $s<a$:

```{figure} ../images/gauss_cil_menor.png
:alt: gauss_cil_menor.png
:width: 260px
s < a.
```

La carga encerrada se calcula por integración, y aprovechando de $\rho$ es constante, $Q_{enc} = \int_V \rho dV = \rho V = \rho H \pi s^2$, y la integral del vector desplazamiento se convierte en el producto del vector desplazamiento radial por la superficie $S_l$ (las tapas no contribuyen a la integral):

$$
D_s H 2 \pi s = \rho H \pi s^2
$$

Para radio $s>a$:

```{figure} ../images/gauss_cil_mayor.png
:alt: gauss_cil_mayor.png
:width: 260px
s > a.
```

La carga es la del trozo de cilindro de carga $Q_{enc} = \int_V \rho dV = \rho V = \rho H \pi a^2$.

$$
D_s H 2 \pi s = \rho H \pi a^2
$$

### Simetría plana

En estos casos, tenemos el problema de que el sistema se extiende hasta el infinito, por lo que no se puede usar como referencia que el campo en el infinito es cero. *Si se puede decir: en la dirección perpendicular a los planos, si la carga neta de todo el sistema es nula, en el infinito el campo vale 0*, como pasa en un condensador plano-paralelo. Vamos a resolver otro ejemplo distinto: La mitad del espacio $z<0$ es de conductor elétrico perfecto (CEP) sin carga, y la otra mitad $z>0$ tiene una densidad de carga $\rho = \rho_0 e^{-z/l}$ en el vacio que decrece exponencialmente, con $\rho_0$ y $l$ constantes.

* El campo eléctrico es perpendicular a los planos de simetría $\vec{E} = E_z \vec{u_z}$. 

* Las superficies con simetría deben ser perpendiculares a los planos de la dirección simétrica. La forma da igual, mientras su superficie total $S_p$ sea constante. El diferencial de superficie es, por tanto, $dS \vec{u_z}$.

* Para cerrar la superficie, usamos caras laterales cuyo diferencial de superficie siempre es perpendicular al campo eléctrico por lo que no contribuyen.

* Una superficie de este tipo debe cerrarse con otra cara paralela, por lo que esta simetría solo sirve para comparar el campo eléctrico en dos regiones. Siempre vamos a tener dos caras paralelas $S_1$ y $S_2$, del mismo tamaño $S_1 = S_2 = S_p$ (*una de ellas se puede hacer tender a infinito si la carga neta es nula y sabemos que en el infinito el campo eléctrico vale 0*). En este ejemplo, podemos usar que en el interior del CEP, el campo es nulo para tener un valor conocido.

* Es importante observar que las dos caras, o tapas, que contribuyen a la integral, tienen diferenciales de superficie en direcciones opuestas, $dS_1 = -dS_2$.

Una de las caras, o tapas, que contribuye a la integral la colocamos en el interior del CEP, mientras que la otra tapa la ponemos en el espacio con densidad de carga:

```{figure} ../images/gauss_pla.png
:alt: gauss_pla.png
:width: 400px
Simetría plana.
```

La cara $S_2$ no contribuye a la integral **en este caso**, por que dentro del CEP, el campo debe ser nulo. La integral de la cara $S_1$ es el producto, y además observamos que con signo positivo, ya que el diferencial de superficie y el campo eléctrico son positivos hacia arriba. La carga encerrada depende de la posición a la que se coloca la cara $S_1$, genéricamente en $z$, ya que no se confunde con otros parámetros del problema.

$$
E_z S_p = \frac{1}{\varepsilon_0} \int_{S_1} \int_0^z\rho_0 e^{-z/l}dzdS_1 = \frac{S_p}{\varepsilon_0} \left( -l \rho_0 e^{-z/l} \right)_0^z = \frac{S_p}{\varepsilon_0} \left( l \rho_0(1 - e^{-z/l} \right)
$$

¡Y no hay más simetrías de Gauss posibles!

### Ejercicios propuestos

Problemas inmediatos de todas las simetrías posibles:

* Carga puntual $q$:

$$
E_r = \frac{q}{4 \pi \varepsilon_0 r^2}
$$

* Esfera hueca de radio $a$ y carga total $Q = \sigma 4 \pi a^2$.

$$
E_r &= \frac{Q}{4 \pi \varepsilon_0 r^2} \quad & r>a \\
E_r &= 0 & r<a
$$

* Esfera de radio $a$ y carga total $Q = \rho \frac{4}{3} \pi a^3$.

$$
E_r &= \frac{Q}{4 \pi \varepsilon_0 r^2} \quad & r>a \\
E_r &= \frac{Q r}{4 \pi \varepsilon_0 a^3} & r<a
$$

* Hilo infinito de densidad lineal de carga $\lambda$.

$$
E_s = \frac{\lambda}{2 \pi \varepsilon_0 s}
$$

* Cilindro infinito hueco de radio $a$ y densidad superficial de carga $\sigma$.

$$
E_s &= \frac{\sigma a}{\varepsilon_0 s} \quad &s>a \\
E_s &=0 \quad &s<a
$$

* Cilindro infinito de radio $a$ y densidad de carga $\rho$.

$$
E_s &= \frac{\rho a^2}{2 \varepsilon_0 s} \quad &s>a \\
E_s &= \frac{\rho s}{2 \varepsilon_0} \quad &s<a
$$

* Plano infinito en $z=0$ de densidad superficial de carga $\sigma$.

$$
E_z &= \frac{\sigma}{2 \varepsilon_0} \quad &z>0 \\
E_z &= -\frac{\sigma}{2 \varepsilon_0} \quad &z<0
$$


## Magnetostática: Ley de Ampere

En magentostática, la ecuación que nos permite aprovechar la simetría es la de la ley de Ampere en forma integral:

$$
\int_C \vec{H} \cdot \vec{dl} = I_{enc}
$$

O en su versión para permeabilidad magnética constante:

$$
\int_C \vec{B} \cdot \vec{dl} = \mu I_{enc}
$$

En el caso del vacio, simplemente usamos la permeabilidad magnética del vacio:

$$
\int_C \vec{B} \cdot \vec{dl} = \mu_0 I_{enc}
$$

Recordemos que:

$$
I_{enc} = \int_S \vec{J} \cdot \vec{dS}
$$

**¡Esta ecuación es válida siempre!** Es decir, cualquier circuito $C$ se puede usar para realizar la integral y sumar la corriente total que atraviesa la superficie delimitada por $C$. *Es fundamental saber elegir el circuito que facilita y permite resolver el problema*:

* La simetría y la regla de la mano derecha nos deben permitir predecir la dirección del campo magnético $\vec{B}$ o del vector $\vec{H}$.

* El circuito debe tener algún lado que recorra la dirección del campo magnético $\vec{B}$ o del vector $\vec{H}$.

* Si el circuito no cierra naturalmente, debemos cerrarlo con segmento perpendiculares a la dirección del campo magnético $\vec{B}$ o del vector $\vec{H}$.

* Debemos poder calcular la corriente que atraviesa el interior del circuito $C$.

En estos problemas la simetría esférica es imposible, pero por otra parte, las simetrías cilíndrica y planar tienen varias direcciones posibles. Veamos:

### Simetría cilíndrica

La simetría cilíndrica tiene dos simetrías:
* En la dirección del eje.
* En la rotación en torno al eje.

Por lo tanto, la corriente con simetría cilíndrica puede tener dos direcciones.
* Corriente a lo largo del eje.
* Corriente alrededor del eje.

El primer caso sería un hilo de corriente. Por ejemplo, supongamos que tenemos el hilo cilindrico de radio $a$ con densidad de corriente $\vec{J}$ uniforme a lo largo del hilo.

```{figure} ../images/hilo.png
:alt: hilo.png
:width: 250px
Hilo de corriente.
```

La regla de la mano derecha nos indica que el campo magnético debe girar en la dirección $\vec{u_\varphi}$ de coordenadas cilíndricas. Por lo tanto, el circuito de Ampere debe seguir la misma dirección, y girar en torno al hilo. La corriente claramente atraviesa el circuito y se puede calcular. Tenemos dos posibilidades, que el radio $r$ sea $r<a$ o $r>a$.

```{figure} ../images/hilo_ampere.png
:alt: hilo_ampere.png
:width: 250px
Hilo de corriente.
```

Con esto, la integral de Ampere es el producto del campo magnético por su longitud, mientras que la corriente total (**siendo $\vec{J}$ constante**) es el producto de la densidad de corriente por el área. Así:

$$
B 2 \pi r &= \mu_0 J \pi r^2 \quad & r<a \\
B 2 \pi r &= \mu_0 J \pi a^2 \quad & r>a
$$

El segundo caso de posible simetría es que la corriente gire en la dirección $\vec{u_\varphi}$, por lo que tenemos un solenoide. Por ejemplo, veamos el solenoide de radio $a$ con $n$ vueltas por unidad de longitud.

```{figure} ../images/solenoide.png
:alt: solenoide.png
:width: 250px
Solenoide.
```

¡El ciruito de Ampere anterior no sirve! La regla de la mano derecha nos dice que el campo magnético dentro del hilo debe ir en la dirección $\vec{u_z}$, por lo que el circuito de Ampere debe tener un lado en esa dirección. Para que no se extienda infinitamente, lo cerramos con lados perpendiculares, y finalmente otro lado vertical, teniendo un marco.

```{figure} ../images/solenoide_ampere.png
:alt: solenoide_ampere.png
:width: 250px
Solenoide.
```

De esta forma, el lado interno contribuye a la integral, mientras que los otros 3 no, ya que fuera, el campo magnético es nulo. 

```{note}
Demostremos que el campo magnético fuera del solenoide es nulo. Si supusieramos que el campo magnético fuera no es nulo, veremos que debe ser constante. Podemos despues hacer que el lado exterior tienda a infinito. Entonces comprobaremos que el valor fuera debe coincidir con el valor en infinito, es decir, nulo.
```

La corriente que atraviesa el circuito también es facil, con el número de vueltas por unidad de longitud.

$$
B_z l = \mu_0 I n l
$$

El campo magnético en el interior del solenoide cilíndrico es por tanto constante. Veamos que si doblamos el solenoide levemente, el campo variará poco. Si el doblez es intenso, el campo magnético se concentra en la parte interna, donde los hilos de corriente se apelotonan.

La manipulación del solenoide nos da una última simetría cilíndrica, el solenoide cerrado con $n$ vueltas por unidad de longitud, que es el que tiene mayor aplicación práctica.

```{figure} ../images/solenoide_cerr.png
:alt: solenoide_cerr.png
:width: 350px
Solenoide cerrado.
```

En este caso, el campo magnético también va por el interior y es nulo fuera, igual que en el solenoide cilíndrico. Mas en concreto, por el interior y por la regla de la mano derecha, en la dirección $-\vec{u_\varphi}$, por lo que el circuito de Ampere debe tener esta dirección. Vemos que la corriente atraviesa el circuito por el interior del solenoide cerrado, y se puede calcular.

```{figure} ../images/solenoide_cerr_ampere.png
:alt: solenoide_cerr_ampere.png
:width: 350px
Solenoide cerrado.
```

$$
B 2 \pi r = \mu_0 I n 2 \pi a
$$

Efectivamente, el campo es mayor en la parte interior del doblez que en la exterior.


### Simetría plana

Solo existe un tipo de simetría plana, la de la corriente en una dirección constante a lo largo y ancho de un plano. Por ejemplo, digamos que la dirección es $\vec{u_x}$ en el plano $z=0$ y la densidad superficial de corriente es $\vec{j} = I n \vec{u_x}$ con $n$ el numero de hilos de corriente por unidad de longitud.

```{figure} ../images/corriente_pla.png
:alt: corriente_pla.png
:width: 350px
plano de corriente.
```

Por la regla de la mano derecha, el campo magnético debe ir en la dirección $\vec{u_y}$ para $z$ negativos y en la dirección $-\vec{u_y}$ para $z$ positivos. Si suponemos que el campo magnético se reparte por igual en los dos lados, podemos decir que ambos son iguales en modulo. El circuito de Ampere debe tener dos lados en la dirección $y$ que se deben cerrar con otros lados en la dirección $z$, perpendicular al campo magnético.

```{figure} ../images/corriente_pla_ampere.png
:alt: corriente_pla_ampere.png
:width: 350px
plano de corriente.
```

$$
B_y 2 l = \mu_0 j l
$$

¡Y no hay más simetrías de Ampere posibles!
