# Propiedades magnéticas de los medios materiales en régimen estático

Hasta ahora, desde el punto de vista magnetostático, hemos utilizados distintos tipos de materiales para estudiar el comportamiento del campo magnético $\vec{B}$:

* **Vacío**: Caracterizado por las ecuaciones de la magnetostática y por la permeabilidad magnética del vacío $\mu_0$.

* **Materiales conductores eléctricos perfectos CEP)**: Caracterizados por la presencia de cargas con movilidad infinita (*con respecto a la escala de tiempos de los problemas que podemos estudiar*). Nos permiten imponer unas condiciones de contorno sobre las ecuaciones de la magnetostática para resolver algunos problemas de forma más sencilla. Recordemos que en el interior, el campo magnético $\vec{B}$ es nulo y al otro lado de la frontera del CEP, debe ser tangencial a la superficie de la frontera.

* **Materiales conductores**: Los caracterizamos por una densidad de corriente $\vec{J}$. Podemos calcular el campo magnético en su entorno mediante la ley de Biot-Savart.

* **Materiales no conductores**: Están constituidos por átomos y moléculas que contienen electrones que describen órbitas, desde el punto de vista clásico, o que pueblan un orbital con un determinado momento angular $\vec{l}$. Además, los electrones y algunos núcleos tienen un espín, con otro momento angular intrínseco. Estos momentos angulares implican unas corrientes que llamamos ligadas y por tanto, un momento magnético microscópico $\vec{m}$. *Se modelan utilizando la permeabilidad magnética del material o su permeabilidad relativa y utilizando la definición de vector $\vec{H}$*. Estas definiciones nos permiten modificar las leyes de la magnetostática para no tener que escribir explícitamente las corrientes ligadas, que quedan modeladas mediante la permeabilidad relativa del material. **Observemos que aún no hemos relacionado la dinámica microscópica de momentos magnéticos microscópicos con el efecto macroscópico**: Este es el objetivo de este tema.

```{note}
El objetivo de este tema es relacionar la respuesta magnética macroscópica de un material con la respuesta magnética microscópica de las corrientes ligadas y de los momentos magnéticos de las partículas del material.
```

También vamos a realizar varios **modelos teóricos** para ciertos conceptos, **y vamos a extraer algunas conclusiones a partir de ellos**.


## Imanación

Vamos a definir la imanación a partir del efecto combinado de los momentos magnéticos microscópicos. *Una configuración con un momento magnético se puede simplificar a todos los efectos por un pequeño circuito de forma arbitraria, o dipolo magnético, con superficie $S$ y por el que pasa una corriente $I$. Se le puede asociar un momento magnético dipolar que, para un circuito plano, es perpendicular a dicho plano, de valor $\vec{m}$*:

$$
\vec{m} = I S \vec{n}
$$

La imanación $\vec{M}$ se obtiene sumando todos los momentos magnéticos en una región de volumen $\Delta V$**:

$$
\vec{M} = \lim_{\Delta V \rightarrow 0} \frac{\sum \vec{m_i}}{\Delta V} = n \left<\vec{m}\right>
$$

Observemos que las unidades de la imanación son $\frac{A}{m}$, las misma que la del vector $\vec{H}$. Esta definición es la más conveniente, porque como también vimos, la imanación de un material se puede sustituir por unas densidades de corrientes ligadas volumétrica $\vec{J_m}$ y superficial $\vec{j_m}$, y resolver los problemas de magnetostática utilizando dichas densidades de corriente. Siendo $\vec{n}$ el vector saliente de la superficie que envuelve un volumen con imanación $\vec{M}$:

$$
\vec{J_m} &= \nabla \times \vec{M} \\
\vec{j_m} &= \vec{M} \times \vec{n}
$$

En muchos materiales, la imanación es proporcional al vector $\vec{H}$, mediante la constante susceptibilidad magnética del material $\chi_m$.

$$
\vec{M} = \chi_m \vec{H}
$$

La definición del vector $\vec{H}$ es la que surge de la siguiente expresión, de tal forma que las ecuaciones de la magnetostática se escriben sólo en función de las corrientes de conducción externas, y las corrientes ligadas $\vec{J_m}$ y $\vec{j_m}$ quedan modeladas a través de la permeabilidad magnética del material o de la permeabilidad relativa.

$$
\vec{B} = \mu_0 \left( \vec{H} + \vec{M} \right) = \mu_0 \left(1 + \chi_m \right) \vec{H} = \mu_0 \mu_r \vec{H} = \mu \vec{H}
$$

```{note}
En todo este recordatorio, la única relación entre lo microscópico y lo macroscópico es la que da la expresión:

$$
\vec{M} =  n \left<\vec{m}\right>
$$
```


## Fuentes del magnetismo microscópico

El magnetismo microscópico puede provenir de dos tipos de fuente:

* **El espín** intrínseco de algunas partículas, como por ejemplo los electrones.

* **Las órbitas** que describen los electrones en torno a los núcleos atómicos.

Vamos a estudiar los dos tipos de magnetismo microscópico poniendo como ejemplo un tipo de material para el que es relevante.

En general, el vector $\vec{H_L}$ local que actúa sobre cada átomo no es el mismo que el que se impone desde el exterior $\vec{H}$, o en términos de campo magnético $\vec{B_L} \neq \vec{B}$. En este sentido podemos desarrollar varios modelos para relacionar el vector $\vec{H_L}$ local y el vector $\vec{H}$ externo.

* **Modelo de gas diluido**: Si suponemos que los dipolos magnéticos son suficientemente independientes y no se afectan mutuamente, el vector $\vec{H_L}$ local es prácticamente el vector $\vec{H}$ externo y $\vec{H_L} \approx \vec{H}$.

* **Modelo de Lorentz**: El material se modela como un conjunto de dipolos orientados en la misma dirección. Para los dipolos cercanos se calcula el vector $\vec{H}$ real que generan, mientras que para los dipolos lejanos, se utilizan las densidades de corriente ligadas. Se define una frontera esférica de radio arbitrario que incluya suficientes dipolos cercanos, llamada la esfera de Lorentz.

* **Otros modelos**: Para conseguir resultados más precisos, en necesario considerar el efecto de los dipolos cercanos, como por ejemplo en el modelo de Ising.


## Materiales diamagnéticos

El tipo de materiales con la respuesta más leve se explican correctamente con el modelo de órbita clásica estudiando cómo se afecta ésta por la presencia de una campo magnético local $\vec{B_L}$. Consideremos una órbita circular de un electrón en torno a un núcleo, que puede ser descrito mediante un momento magnético.

```{figure} ../images/mat_11.png
:alt: mat_11.png
:width: 250px
Órbita clásica sin campo magnético
```

```{note}
En una vuelta completa, la corriente $I$ es simplemente la carga dividida entre el tiempo:

$$
I = \frac{-e}{T} = -\frac{ev}{2\pi R}
$$

Mientras que la superficie de la trayectoria circular es simplemente $S = \pi R^2$. Por lo tanto, el momento magnético de esta trayectoria es:

$$
m = -\frac{evR}{2}
$$

Es interesante calcular, por otra parte, el momento angular y relacionarlo, con el momento dipolar magnético:

$$
l = Rp = m_evR = - \frac{2 m_e}{e} m
$$

O a la inversa:

$$
m = -\frac{e}{2m_e} l
$$
```

La fuerza $F_e$ que le hace girar será eléctrica y se iguala a la fuerza centrípeta de una órbita con velocidad $v_0$ y radio $R$.

$$
F_e = m_e \frac{v_0^2}{R}
$$


Sin cambiar el radio de la órbita, la fuerza eléctrica es la misma. Si añadimos un campo magnético local $\vec{B_L}$, la fuerza magnética de una trayectoria circular es conocida, $F_B = evB_L$, y la velocidad debe incrementar para dar cuenta de la mayor fuerza central. 

$$
F_e + F_B = m_e \frac{v^2}{R}
$$

```{figure} ../images/mat_12.png
:alt: mat_12.png
:width: 250px
Órbita clásica con campo magnético
```

Por lo tanto, despejando $\Delta(v^2) = v^2 - v_0^2$.

$$
m_e \Delta(v^2) = evR B_L
$$

Si el cambio en la velocidad es suficientemente pequeño, $\Delta(v^2) = 2v\Delta v$ y por tanto:

$$
\Delta v = \frac{e R B_L}{2 m_e}
$$

Finalmente, podemos obtener el cambio en el momento dipolar magnético de la órbita del electrón:

$$
\Delta m = -\frac{e}{2 m_e}\Delta l = -\frac{e}{2 m_e} \left( m_e R \Delta v \right) = -\frac{e}{2 m_e} (m_e R) \frac{e R B_L}{2 m_e} = -\frac{e^2R^2}{4 m_e}B_L
$$

En un material con una densidad de átomos $n$, si cada átomo tiene $Z$ electrones con distintas órbitas, tenemos que sumar, para la imanación, la modificación en el momento de cada órbita para todos los electrones.

$$
M = n \sum_{i=1}^Z \Delta m_i = n \left( \frac{e^2}{4 m_e} \sum_{i=1}^Z r_i^2 \right) B_L = n \left( \frac{e^2 \mu_0}{4 m_e} \sum_{i=1}^Z r_i^2 \right) H_L
$$

El factor que multiplica al vector $H_L$, si realizamos la operación, da un valor pequeño por lo que podemos suponer que el efecto de las órbitas cercana es nulo y utilizar la hipótesis de gas diluido, $H_L \approx H$ en un material diamagnético. Por lo tanto, ya tenemos la susceptibilidad magnética $\chi_m$:

$$
M \approx n \left( \frac{e^2 \mu_0}{4m_e} \sum_{i=1}^Z r_i^2 \right) H = \chi_m H
$$

* La susceptibilidad es negativa.

* La susceptibilidad es del orden de $-10^{-5}$.


## Materiales paramagnéticos

En los materiales paramagnéticos, la respuesta magnética es más intensa. Se da en los materiales en los que la orientación de los momentos dipolares magnéticos puede variar. Cuando se impone un campo magnético externo, las distintas orientaciones pasan a tener distinta energía. **La imanación máxima que pueden tener es precisamente la de todos los momentos dipolares magnéticos orientados en la misma dirección**:

$$
\vec{M_{max}} = n \vec{m_0}
$$

En este caso, la energía de todos los momentos dipolares magnéticos es mínima. Sin embargo, si la sustancia tiene una cierta temperatura, los dipolos magnéticos intentarán orientarse debido al efecto del campo magnético local $\vec{B_L}$, pero las colisiones y vibraciones aleatorias producirán una orientación aleatoria. **Debemos utilizar los principios de la Física estadística para obtener la situación de equilibrio, es decir, utilizamos la distribución de Maxwell-Boltzmann**. Para ello, necesitamos la energía de cada dipolo magnético con orientación arbitraria, $\vec{m}$ con $m = m_0$  :

$$
U_m = -\vec{m} \cdot \vec{B_L}
$$

Vamos a estudiar dos posibilidades:

````{note}
**Materiales en los que el momento dipolar magnético puede tener cualquier orientación.**

El número de momentos dipolares $\vec{m}$ con una orientación determinada en un determinado volumen $\Delta V$ es proporcional a:

$$
n(\vec{m}) \propto \exp \left(-\frac{U_m}{k_BT} \right) = \exp \left(\frac{\vec{m} \cdot \vec{B_L}}{k_BT} \right)
$$

Si elegimos la orientación del campo magnético local en la dirección $z$:

$$
n(\vec{m}) \propto \exp \left(\frac{m_0 B_L \cos\theta}{k_BT} \right)
$$

Por lo tanto el problema se reduce a plantear, en el espacio $(\theta, \phi)$ de esféricas, la integral de la densidad de partículas en cada dirección multiplicada por el momento magnético en esa dirección.

$$
\left<M_z\right> = \left< n m_0 \cos\theta \right> = n m_0 \left<\cos\theta \right>
$$

Este problema tiene como solución la conocida como función de Langevin $\mathcal{L}(y) = \coth{y} - \frac{1}{y}$:

$$
\left<M_z\right> = n m_0 \mathcal{L}\left( \frac{m_0 B_L}{k_BT} \right)
$$

En el límite para bajas temperaturas, la función de Langevin tiende a 1, por lo que efectivamente, predice que todos los momentos dipolares magnéticos se orientan en la dirección del campo magnético local $\vec{B_L}$. **Para temperaturas suficientemente altas**, la podemos aproximar por $\mathcal{L}(y) = \coth{y} - \frac{1}{y} \approx \frac{y}{3}$. De hecho, **en la mayoría de las situaciones, podremos utilizar esta aproximación**.


```{figure} ../images/mat_4.png
:alt: mat_4.png
:width: 400px
Función de Langevin.
```

````

````{note}
**Materiales en los que el momento dipolar magnético está cuantizado.**

Como ejemplo, vamos a utilizar un sistema de espines electrónicos. El momento dipolar magnético asociado al espín del electrón es:

$$
U_{\uparrow,\downarrow} = \mp m_B B_L
$$

El número de electrones con espín en cada dirección en un volumen $\Delta V$, es proporcional a:

$$
n_{\uparrow,\downarrow} \propto \exp \frac{-U_{\uparrow,\downarrow}}{k_BT} \exp \frac{\pm m_B B_L}{k_BT}
$$

Claramente, $n = n_\uparrow + n_\downarrow$, por lo que la expresión de la imanación:

$$
\left<M_z\right> = n_\uparrow m_B - n_\downarrow m_B = n \frac{n_\uparrow m_B - n_\downarrow m_B}{n_\uparrow + n_\downarrow} = n m_B \frac{\exp \frac{+m_B B_L}{k_BT}-\exp \frac{-m_B B_L}{k_BT}}{\exp \frac{+m_B B_L}{k_BT}+\exp \frac{-m_B B_L}{k_BT}} = n m_B \tanh \frac{m_B B_L}{k_BT}
$$

En el límite para bajas temperaturas, la tangente hiperbólica tiende a 1, por lo que efectivamente, predice que todos los momentos dipolares magnéticos se orientan en la dirección del campo magnético local $\vec{B_L}$. **Para temperaturas suficientemente altas**, la podemos aproximar por $\tanh(y) \approx y$. De hecho, **en la mayoría de las situaciones, podremos utilizar esta aproximación**.
````

En ambos casos obtenemos una expresión para $\chi_m$, para la aproximación para temperaturas suficientemente altas. Sustituyendo $\vec{B_L} = \mu_0 \vec{H_L}$:

* Para materiales en los que el momento dipolar magnético puede tener cualquier orientación.

$$
M = \frac{n m_0^2 \mu_0}{3 k_BT} H_L \propto \frac{1}{T}
$$

* Para sistemas de espines electrónicos.

$$
M = \frac{n m_B^2 \mu_0}{k_BT} H_L \propto \frac{1}{T}
$$

En ambos casos, es inversamente proporcional a la temperatura $T$. Para terminar los modelos, necesitamos relacionar el vector $\vec{H_L}$ local con el vector $\vec{H}$ externo. En sustancias paramagnéticas, en algunas ocasiones, la aproximación del modelo de gas diluido da resultados buenos, pero en general no se puede despreciar el efecto de los dipolos magnéticos próximos.

$$
\vec{H_L} = \vec{H} + \gamma \vec{M}
$$

Donde $\gamma$ depende del modelo (*de Lorentz, de Ising...*). En ambos casos, el resultado es el mismo, la ley conocida como Ley de Curie-Weiss, aunque obviamente, el cálculo de la temperatura crítica $T_c$ será diferente en función del modelo:

$$
\chi_m = \frac{C}{T - T_c}
$$


# Dominios ferromagnéticos y paredes de Bloch

Por debajo de la temperatura crítica, los momentos dipolares magnéticos tienen tendencia a alinearse en la misma dirección. Si el material es suficientemente homogéneo, es posible que el material entero se oriente en la misma dirección (*como los imanes naturales*), pero en general, **se van a formar dominios ferromagnéticos con orientaciones aleatorias**. De hecho, el material entero puede tener una imanación neta nula, aunque tenga imanaciones locales.

```{figure} ../images/mat_13.png
:alt: mat_13.png
:width: 400px
Dominios ferromagnéticos.
```

La frontera entre distintos dominios ferromagnéticos son **las paredes de Bloch: Zonas de transición que pueden ser de tan sólo unos cientos de átomos en los que la imanación de las distintas zonas se adapta**. Imaginemos cómo se forman, empezando con el material suficientemente caliente como para no tener imanación espontánea. Cuando se enfría y se acerca a la temperatura crítica, los dipolos magnéticos próximos empiezan a intentar orientarse en la misma dirección. **En distintas zonas del material, pueden empezar a orientarse en direcciones distintas, y estas regiones van creciendo y forman los dominios ferromagnéticos**. Cuando dos dominios ferromagnéticos entran en contacto, se forma la zona de transición que encuentra el equilibrio entre los dominios que intentan crecer y los propios dipolos que prefieren estar alineados.

```{figure} ../images/mat_14.png
:alt: mat_14.png
:width: 400px
Paredes de Bloch.
```

**Las paredes de Bloch no son fijas**. Pueden desplazarse por el efecto de un vector $\vec{H}$ externo.

* **Si el vector $\vec{H}$ externo es muy intenso**, el material entero se magnetiza en la misma dirección. Dicha magnetización en general se mantiene cuando el campo magnético externo desaparece.
* **Si el vector $\vec{H}$ externo no es tan intenso**, algunas regiones que estén orientadas con dicho campo magnético, tienen una energía menor que otras, y por lo tanto tendrán tendencia a crecer. Dependiendo del material, puede ser que cada dominio ferromagnético se oriente poco a poco, o puede ser que las paredes de Bloch se desplacen para que los dominios ferromagnéticos con menor energía crezcan.
